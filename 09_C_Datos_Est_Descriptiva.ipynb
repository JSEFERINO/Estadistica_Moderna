## **d. Descriptores numerícos para una columna de la Data**


```python
## **d. Descriptores numerícos  de la Data**
df_datos2["EDAD"].describe()
```

## **e. Editando a nuestros descriptores numéricos - La Media o Promedio**

```python
## **e. Editando a nuestros descriptores numéricos - La Media o Promedio**
print("La media de la variable EDAD es igual a", df_datos2["EDAD"].mean())
```

## **f. Editando a nuestros descriptores numéricos - La Mediana**

```python
## **f. Editando a nuestros descriptores numéricos - La Mediana**
print("La mediana de la variable EDAD es igual a", df_datos2["EDAD"].median())
```

## **g. Editando a nuestros descriptores numéricos - La Moda**

```python
## **g. Editando a nuestros descriptores numéricos - La Moda**
print("La moda de la variable EDAD es igual a", df_datos2["EDAD"].mode())
```

## **h. Editando a nuestros descriptores numéricos - La Desvacion estandard**

```python
## **h. Editando a nuestros descriptores numéricos - La Desvacion estandard**
print("La desviación estandard de la variable EDAD es igual a", df_datos2["EDAD"].std())
```

## **i. Editando a nuestros descriptores numéricos - Percentiles**

```python
## **i. Editando a nuestros descriptores numéricos - Percentiles**
print("El 75% de los estudiantes tienen edades por debajo de ", np.percentile(df_datos2["EDAD"], 75))
```

```python
## **i. Editando a nuestros descriptores numéricos - Percentiles**
print("El 85% de los estudiantes tienen edades por debajo de ", np.percentile(df_datos2["EDAD"], 85))
```

## **j. Editando a nuestros descriptores numéricos - Máximos y mínimos**

```python
## **j. Editando a nuestros descriptores numéricos - Máximos y mínimos**
print("Mis estudiantes tienen edades entre ",'(' ,df_datos2["EDAD"].min(),',' ,df_datos2["EDAD"].max(),')' )
```
