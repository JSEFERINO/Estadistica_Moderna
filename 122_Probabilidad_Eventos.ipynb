# **2.1.0. Breve historia de la probabilidad - por Chat GPT**
- La historia de la probabilidad es fascinante y abarca varios siglos de desarrollo. Aquí te presento un breve recorrido por los hitos más destacados en la evolución de la probabilidad:

## **2.1.1. Orígenes Antiguos**
- Los primeros indicios de conceptos relacionados con la probabilidad se remontan a civilizaciones antiguas como los egipcios y los romanos, quienes utilizaban dados y otros juegos de azar. Sin embargo, estos enfoques iniciales estaban más relacionados con la práctica que con la teoría.

## **2.1.2. Cartas y Correspondencia:**

+ En el siglo XVI, los matemáticos italianos *Gerolamo Cardano* y *Girolamo Saccheri* realizaron contribuciones significativas al estudio de juegos de cartas y probabilidad. Cardano, en particular, escribió un libro titulado ***Liber de Ludo Aleae (El libro sobre los juegos de azar)***, en el que discutió conceptos de probabilidad.

## **2.1.3. Pascal y Fermat:**
+ En el siglo XVII, *Blaise Pascal* y *Pierre de Fermat* colaboraron en la resolución de problemas de juegos de azar que involucraban apuestas y probabilidades. Sus correspondencias dieron lugar a la formulación de lo que hoy se conoce como el Triángulo de Pascal y la Regla de Fermat-Pascal, herramientas fundamentales en la teoría de la probabilidad.

## **2.1.4. Laplace y la Probabilidad Clásica:**
+ A fines del siglo XVIII, ***Pierre-Simon Laplace***, un matemático y físico francés, contribuyó significativamente al desarrollo de la probabilidad clásica. Propuso la idea de que la probabilidad podría entenderse como una medida de la incertidumbre y formuló el Principio de Indiferencia, que se utilizó para asignar probabilidades a eventos.

## **2.1.5. Teoría de la Probabilidad Formal:**
+  Durante el siglo XIX, la probabilidad comenzó a desarrollarse como una teoría formal, en parte gracias a los trabajos de matemáticos como ***Andrei Markov*** y ***Aleksandr Lyapunov***. Surgieron teoremas y conceptos fundamentales, como la ***Ley de los Grandes Números*** y el ***Teorema Central del Límite***, que sentaron las bases de la estadística moderna.


## **2.1.6. Siglo XX y la Revolución Estadística:**
-  El siglo XX presenció una explosión en la teoría de la probabilidad y la estadística. Figuras destacadas como ***Ronald A. Fisher, Jerzy Neyman, Abraham Wald*** y otros contribuyeron significativamente al desarrollo de la estadística moderna y su aplicación en campos tan diversos como la ***biología, la economía y la ciencia de datos***.

## **2.1.7. Probabilidad en la Era de la Computación:**
- Con el avance de la informática, la ***probabilidad*** se convirtió en una herramienta esencial en el _análisis de datos y la toma de decisiones_. La simulación Monte Carlo y técnicas de probabilidad bayesiana se volvieron cruciales en campos como la inteligencia artificial y la ciencia de datos.

**En resumen, la historia de la probabilidad es una narrativa de progreso continuo a lo largo de los siglos, desde sus raíces en los juegos de azar hasta su papel central en la ciencia y la toma de decisiones en la era moderna. La probabilidad es ahora una disciplina esencial en matemáticas y estadísticas, con aplicaciones que abarcan una amplia gama de campos y sectores de la sociedad.**

### $\blacksquare$ **Ejemplo:** **Simulación de lanzamiento de un dado.**

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import numpy as np

# Función para lanzar el dado
def lanzar_dado():
    return np.random.randint(1, 7)

# Inicialización de variables para contar los resultados
resultados = {str(i): 0 for i in range(1, 7)}
historial_resultados = []

# Función para actualizar los resultados, la tabla y el gráfico en cada fotograma de la animación
def actualizar(i):
    resultado = lanzar_dado()
    resultados[str(resultado)] += 1
    historial_resultados.append(resultado)

    ejes[0].clear()
    ejes[0].text(0.5, 0.5, str(resultado), fontsize=50, ha='center', va='center')

    ejes[1].clear()
    ejes[1].bar(resultados.keys(), resultados.values(), color='blue')
    ejes[1].set_ylim(0, max(resultados.values()) + 1)

    ejes[2].clear()
    ejes[2].axis('off')
    tabla = ejes[2].table(cellText=[historial_resultados], colLabels=list(range(1, len(historial_resultados)+1)), cellLoc='center', loc='center')
    tabla.auto_set_font_size()
    tabla.set_fontsize(10)

# Configuración inicial del gráfico
fig, ejes = plt.subplots(1, 3, figsize=(10, 5))

# Configuración de la animación
ani = animation.FuncAnimation(fig, actualizar, frames=29, interval=500, repeat=False)

# Muestra la animación en el notebook
HTML(ani.to_html5_video())
```

### $\blacksquare$ **Ejemplo:**  **Simulación de los resultados del Baloto**


```python
import random

def simular_baloto():
    numeros_principales = random.sample(range(1, 44), 5)
    balota_adicional = random.randint(1, 16)
    return numeros_principales, balota_adicional

# Ejemplo de simulación
print(simular_baloto())
```

```python
import matplotlib.pyplot as plt

def graficar_resultados(numeros_principales, balota_adicional):
    fig, ax = plt.subplots()

    # Graficar los números principales
    for i, numero in enumerate(numeros_principales):
        ax.text(i, 0.5, str(numero), fontsize=20, ha='center')

    # Graficar la balota adicional
    ax.text(len(numeros_principales), 0.5, str(balota_adicional), fontsize=20, ha='center', color='red')

    # Configuración de la gráfica
    ax.set_xlim(-1, len(numeros_principales) + 1)
    ax.set_ylim(0, 1)
    ax.axis('off')

    plt.show()

# Ejemplo de gráfico
numeros_principales, balota_adicional = simular_baloto()
graficar_resultados(numeros_principales, balota_adicional)
```

```python
import matplotlib.animation as animation

def crear_animacion(resultados):
    fig, ax = plt.subplots()

    def actualizar(frame):
        ax.clear()
        numeros_principales, balota_adicional = resultados[frame]
        for i, numero in enumerate(numeros_principales):
            ax.text(i, 0.5, str(numero), fontsize=20, ha='center')
        ax.text(len(numeros_principales), 0.5, str(balota_adicional), fontsize=20, ha='center', color='red')
        ax.set_xlim(-1, len(numeros_principales) + 1)
        ax.set_ylim(0, 1)
        ax.axis('off')

    ani = animation.FuncAnimation(fig, actualizar, frames=len(resultados), repeat=False)
    return ani

# Generar múltiples resultados para la animación
resultados = [simular_baloto() for _ in range(20)]
ani = crear_animacion(resultados)


# Muestra la animación en el notebook
HTML(ani.to_html5_video())

```
