## **a1. Hallar un polinomio de grado _n_ que aproxime a la función _seno_**

```python
## **a1. Hallar un polinomio de grado _n_ que aproxime a la función _seno_**
# Vamos a crear este conjunto de datos

x = np.linspace(0,4*np.pi,10)

# Queremos que los datos se ajusten a la función sin(x)
y = np.sin(x)

# Los coeficientes del Polinomio de grado n
n = int(input('Ingrese el grado del polinomio de ajuste a los datos '))

# Creamos nuestro polinomio de acuerdo al grado
p = np.polyfit(x,y,n);

print(f'Los coeficientes del polinomio de grado {n} es P(x) = {p}')

# Vamos a graficar el polinomio ajustado

# El recorrido sea el mismo arriba

x1 = np.linspace(0,4*np.pi,1000)

# Este es el polinomio ajustado
pol_ajustado = np.polyval(p,x1);

plt.plot(x,y,'o')
plt.plot(x1,pol_ajustado)
```


## **a3. Primeros Polinomios que se aproximan a la función $seno$**

```python

## **a3. Primeros Polinomios que se aproximan a la función $seno$**
# Vamos a crear este conjunto de datos

x = np.linspace(0,4*np.pi,10)

# Definimos el polinomio
def fx (x1, coef):
    fx = 0
    n = len(coef) - 1
    for p in coef:
        fx = fx + p*x1**n
        n = n - 1
    return fx

# Queremos que los datos se ajusten a la función sin(x)
y = np.sin(x)

for i in range(0,8):
  coef = np.polyfit(x,y,i);
  print(f'Los coeficientes del polinomio de grado {i} es P(x) = {coef}')
  x1 = np.linspace(0,4*np.pi,1000)
  y1 = fx(x1,coef);
  plt.plot(x,y,'o')
  plt.plot(x1,y1,"--",linewidth=3,color='orange')
  plt.show()
```


## **a3. Primeros Polinomios que se aproximan a la función $y = 1./(1+x)$**

```python

## **a3. Primeros Polinomios que se aproximan a la función $y = 1./(1+x)$**
# Vamos a crear este conjunto de datos

x = np.linspace(0,4*np.pi,10)

# Definimos el polinomio
def fx (x1, coef):
    fx = 0
    n = len(coef) - 1
    for p in coef:
        fx = fx + p*x1**n
        n = n - 1
    return fx

# Queremos que los datos se ajusten a la función sin(x)
y = 1./(1+x)

for i in range(0,10):
  coef = np.polyfit(x,y,i);
  print(f'Los coeficientes del polinomio de grado {i} es P(x) = {coef}')
  x1 = np.linspace(0,4*np.pi,1000)
  y1 = fx(x1,coef);
  plt.plot(x,y,'o')
  plt.plot(x1,y1,"--",linewidth=3,color='orange')
  plt.show()
```

## **a4. Primeros Polinomios que se aproximan a la función $seno$ y valor predicho**

```python
## **a4. Primeros Polinomios que se aproximan a la función $seno$ y valor predicho**
# Vamos a crear este conjunto de datos

x = np.linspace(0,4*np.pi,10)

# Definimos el polinomio
def fx (x1, coef):
    fx = 0
    n = len(coef) - 1
    for p in coef:
        fx = fx + p*x1**n
        n = n - 1
    return fx

# Queremos que los datos se ajusten a la función sin(x)
y = np.sin(x)
v_estimar = 12


for i in range(0,10):
  coef = np.polyfit(x,y,i)
  p = np.polyval(coef, v_estimar)

  print(f'Los coeficientes del polinomio de grado {i} es P(x) = {coef}')
  print(f"para grado {i} la predicción es {p}")

  x1 = np.linspace(0, v_estimar + 1, 1000)
  y1 = fx(x1, coef) # funcion
  plt.figure(figsize=[8,6])
  plt.title("Predicción " + str(i))

  plt.scatter(x,y,s=120,c='blueviolet')
  plt.plot(x1,y1,"--",linewidth=3,color='orange')
  plt.scatter(v_estimar,p,s=200,c='red')
  plt.yticks(range(-1,1,20))
  plt.grid("on")
  ax=plt.gca()
  ax.set_xlabel("$x$")
  ax.set_ylabel("$sin(x)$")
  #plt.savefig("img" + str(i)+".jpg", dpi=600)
  plt.show()
```

```python
## **a4. Primeros Polinomios que se aproximan a la función $y = 1./(1+x)$ y valor predicho**
# Vamos a crear este conjunto de datos

x = np.linspace(0,5,10)

# Definimos el polinomio
def fx (x1, coef):
    fx = 0
    n = len(coef) - 1
    for p in coef:
        fx = fx + p*x1**n
        n = n - 1
    return fx

# Queremos que los datos se ajusten a la función sin(x)
y = 1./(1+x)
v_estimar = 4


for i in range(0,10):
  coef = np.polyfit(x,y,i)
  p = np.polyval(coef, v_estimar)

  print(f'Los coeficientes del polinomio de grado {i} es P(x) = {coef}')
  print(f"para grado {i} la predicción es {p}")

  x1 = np.linspace(0, v_estimar + 1, 1000)
  y1 = fx(x1, coef) # funcion
  plt.figure(figsize=[8,6])
  plt.title("Predicción " + str(i))

  plt.scatter(x,y,s=120,c='blueviolet')
  plt.plot(x1,y1,"--",linewidth=3,color='orange')
  plt.scatter(v_estimar,p,s=200,c='red')
  plt.yticks(range(0,1,20))
  plt.grid("on")
  ax=plt.gca()
  ax.set_xlabel("$x$")
  ax.set_ylabel("$sin(x)$")
  #plt.savefig("img" + str(i)+".jpg", dpi=600)
  plt.show()
```

# **Cuadrado Medio del Error CME para elegir la mejor Predicción**

```python
# **Cuadrado Medio del Error CME para elegir la mejor distribución**
x=np.array([1980,1985,1990,1995,2000,2010,2015,2020]) # Lustros de observación
y=np.array([106.46,123.08,132.12,152.27,180.67,205.05,227.23,249.46]) # Producción de Leche en millones de litros

valor_est = 2023

grado = np.arange(0, 10 + 1, 1) #de 0 a 100 de 1 en 1
aproxi = np.array([])
y_pred_vec = np.array([])

for i in grado:
    coef = np.polyfit(x,y,i)
    p = np.polyval(coef, valor_est)
    aproxi = np.append(aproxi, p)

    # para la MSE
    y_pred_vec = np.array([])
    for j in x: # para la MSE
        y_pred = np.polyval(coef, j)
        y_pred_vec = np.append(y_pred_vec,y_pred)
    #print(f"los y: {y}")
    #print(f"los y_aprox: {y_pred_vec}")
    CME = (sum( (y - y_pred_vec)**2 ))/len(y)
    print(f"Para grado {i} El CME es: {CME}")
plt.figure(figsize=[10,8])
plt.title("Grado del polinomio vs predicción")
plt.plot(grado,aproxi,"--",linewidth=3,color='red')
plt.grid("on")
plt.show()
```
