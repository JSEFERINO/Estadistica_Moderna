```python
from PIL import Image

# Abre la imagen
imagen = Image.open("/content/Fondo TEAMS aprobado por rectoría V052024.png")

# Voltea horizontalmente la imagen
imagen_volteada = imagen.transpose(Image.FLIP_LEFT_RIGHT)

# Guarda la imagen volteada
imagen_volteada.save("imagen_volteada.jpg")

# Cierra la imagen original
imagen.close()
```

```python
from PIL import Image

# Abre la imagen
imagen = Image.open("/content/Fondo TEAMS aprobado por rectoría V052024.png")

# Voltea horizontalmente la imagen
imagen_volteada = imagen.transpose(Image.FLIP_LEFT_RIGHT)

# Guarda la imagen volteada
imagen_volteada.save("imagen_volteada.png")

# Cierra la imagen original
imagen.close()
```

```python
import cv2

# Carga la imagen
imagen = cv2.imread("/content/Fondo TEAMS aprobado por rectoría V052024.png")

# Voltea la imagen horizontalmente
imagen_volteada = cv2.flip(imagen, 1)

# Muestra la imagen volteada
cv2.imshow('Imagen Volteada', imagen_volteada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Load the image
imagen = cv2.imread("/content/Fondo TEAMS aprobado por rectoría V052024.png")

# Flip the image horizontally
imagen_volteada = cv2.flip(imagen, 1)

# Display the flipped image
cv2_imshow(imagen_volteada)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
