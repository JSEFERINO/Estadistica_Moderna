# **184. Distribuciones continuas - La Distribución Beta**
# **a. Generalidades de la Distribución Beta**
+ **Wikipedia**. La distribución beta es una familia de distribuciones continuas de probabilidad definidas en el intervalo $(0,1)$
parametrizada por dos parámetros positivos de forma, denotados por
$\alpha$  y $\beta$ , que aparecen como exponentes de la variable aleatoria y controlan la forma de la distribución. La generalización de esta distribución a varias variables es conocida como la distribución de Dirichlet..

+ **Ejemplos para la distribución Beta**: Veamos algunos casos donde esta distribución es aplicable:

 -La distribución beta se usa generalmente para estudiar las variaciones, a través de varias muestras.

 - De un porcentaje que representa algún fenómeno, por ejemplo, el tiempo diario que la gente dedica a mirar televisión.
 - La distribución beta suele utilizarse para modelar la distribución de estadísticos de orden (por ejemplo, el estadístico de orden késimo de una muestra de variables n uniformes (0, 1) tiene una distribución $beta (k, n + 1 – k))$
 - para modelar eventos que se definen por valos mínimos y máximos.
 - Una distribución Beta es una forma versátil de representar resultados para porcentajes o proporciones. Por ejemplo, ¿cuál es la probabilidad de que un candidato deshonesto gane las próximas elecciones presidenciales? Se podría pensar que la probabilidad es 0,2. Tu amigo podría pensar que es 0,15. La distribución beta te da una manera de describir esto.

# **b. La Función Beta**
- **Definición** $$
{B(α,β)} = \int_0^{1} x^{α-1}(1-x)^{β-1}dx$$
- **La función tiene la forma de convolución.** $$
f(t)*g(t) = \int_0^{t} f(x)g(t-x)dx$$
- **Si en la función beta cambiamos $1$ por $t$ y la reescribimos como una convolución se tiene**: $$
t^{α-1}*t^{β-1} = \int_0^{t} x^{α-1}(t-x)^{β-1}dx$$
- **Usaremos el Teorema de la Transformada inversa de Laplace que dice**: $$ℒ^{-1}[F(s)G(s)]=ℒ^{-1}[F(s)]*ℒ^{-1}[G(s)]$$
+ **También sabemos que**: $$ℒ[t^α]=\frac{Γ(α+1)}{s^{α+1}}$$
+ **De manera que:** $$t^α=ℒ^{-1}[\frac{Γ(α+1)}{s^{α+1}}]$$
+ **Replanteando la integral**
$$t^{α-1}*t^{β-1} = \int_0^{t} x^{α-1}(t-x)^{β-1}dx$$
+ **Nos queda como:**

$$ℒ^{-1}[\frac{Γ(α)}{s^{α}}]*ℒ^{-1}[\frac{Γ(β)}{s^{β}}] = \int_0^{t} x^{α-1}(t-x)^{β-1}dx$$

- **Usaremos el Teorema**
$$ℒ^{-1}[\frac{Γ(α)}{s^{α}}\frac{Γ(β)}{s^{β}}] = \int_0^{t} x^{α-1}(t-x)^{β-1}dx$$
- **Por la Linealidad**
$$Γ(α)Γ(β)ℒ^{-1}[\frac{1}{s^{α+β}}] = \int_0^{t} x^{α-1}(t-x)^{β-1}dx$$

- **Multiplicando y dividiendo por $Γ(α+β)$**
$$\frac{Γ(α)Γ(β)}{Γ(α+β)}ℒ^{-1}[\frac{Γ(α+β)}{s^{α+β}}] = \int_0^{t} x^{α-1}(t-x)^{β-1}dx$$

+ **Aplicando el resultado** $$t^a=ℒ^{-1}[\frac{Γ(a+1)}{s^{a+1}}]$$ **cambiando** $a = α+β-1$

+ **Se tiene al final que**

$$\frac{Γ(α)Γ(β)}{Γ(α+β)}t^{α+β-1} = \int_0^{t} x^{α-1}(t-x)^{β-1}dx$$
+ **Evaluando $t=1$ se tiene la función Beta**

 $$
{B(α,β)} = \int_0^{1} x^{α-1}(1-x)^{β-1}dx=\frac{\Gamma(α)\Gamma(β)}{\Gamma(α+β)}
$$

# **c. Propiedades de la función Beta**
+ $B(α,β)=B(β,α)$
+ $B(α,1)=\frac{\Gamma(α)\Gamma(1)}{\Gamma(α+1)}=\frac{\Gamma(α)(1)}{α\Gamma(α)}=\frac{1}{α}$

+ $B(α+1,β)=\frac{\Gamma(α+1)\Gamma(β)}{\Gamma(α+β+1)}=\frac{α\Gamma(α)\Gamma(β)}{(α+β)\Gamma(α+β)}=\frac{α}{(α+β)}B(α,β)$


# **d. Distribuciones continuas - La Distribución Beta**

+ **Función de Densidad**. Si $X$ es una variable aleatoria continua, se dice que $X$ tiene una distribución Beta con parámetros ${\displaystyle \alpha ,β >0}$ y escribimos $X∼Beta(α,β)$ si su función de densidad es

$$f(x;\alpha, \beta)=\frac{1}{B(α,β)}x^{α-1}(1-x)^{β-1}, \ \ 0<x<1$$

+ Tambien $$E(X)=\mu=\frac{α}{α+β},  \ \ \ \ V(X)=σ^2=\frac{αβ}{(α+β)^2(α+β+1)}$$




# **e. Distribución Beta (4 parámetros)**.
- Uso común: modelo para una variable con ambos límites inferior y superior. Frecuentemente usada como una distribución previa para análisis Bayesiano
- Rango de $X: a \leq X \leq b$

- Función densidad de probabilidad

$${\displaystyle f_{X}(x)={\frac {(x-a)^{\alpha -1}(b-x)^{\beta -1}}{\mathrm {B} (\alpha ,\beta )(b-a)^{\alpha +\beta-1}}}}$$
- Media: La media de la variable aleatoria $X$ es

$${\displaystyle {\text{E}}[X]={a+\frac {b\alpha }{\alpha +\beta }}}$$

- La varianza de $x$ es $$
{\displaystyle {\text{Var}}(X)={\frac {\alpha \beta(b-a)^2 }{(\alpha +\beta +1)(\alpha +\beta )^{2}}}}$$








# **f. Veamos que es una función de densidad**
+ **Sabemos que**
$$f(x;\alpha, \beta)=\frac{1}{B(α,β)}x^{α-1}(1-x)^{β-1}, \ \ 0<x<1$$
+ $$ ∫_{-∞}^{∞}f(x)dx= ∫_{0}^{1}\frac{1}{B(α,β)}x^{α-1}(1-x)^{β-1}dx$$
+ $$ ∫_{-∞}^{∞}f(x)dx= \frac{1}{B(α,β)} ∫_{0}^{1}x^{α-1}(1-x)^{β-1}dx$$
+ **Usando la definición de la función beta se tiene**$$ ∫_{-∞}^{∞}f(x)dx= \frac{1}{B(α,β)} B(α,β)=1$$


# **g. Valor Esperado**
+ $$E(X)=μ= ∫_{-∞}^{∞}xf(x)dx= ∫_{0}^{1}x\frac{1}{B(α,β)}x^{α-1}(1-x)^{β-1}dx$$

+ **Reorganzando la integral**
+ $$E(X)= \frac{1}{B(α,β)}∫_{0}^{1}x^{(α+1)-1}(1-x)^{β-1}dx$$

+ **Por las propiedades y la definición de Beta**
+ $$E(X)= \frac{1}{B(α,β)}∫_{0}^{1}x^{(α+1)-1}(1-x)^{β-1}dx=\frac{1}{B(α,β)}B(α+1,β)$$
+ **Reemplazando se tiene que**

 $B(α+1,β)=\frac{α}{(α+β)}B(α,β)$

 $$E(X)=\frac{α}{α+β}$$




# **h. Varianza**
+ $$E(X^2)= ∫_{-∞}^{∞}xf(x)dx= ∫_{0}^{1}x^2\frac{1}{B(α,β)}x^{α-1}(1-x)^{β-1}dx$$

+ **Reorganzando la integral**
+ $$E(X^2)= \frac{1}{B(α,β)}∫_{0}^{1}x^{(α+2)-1}(1-x)^{β-1}dx$$

+ **Por las propiedades y la definición de Beta**
+ $$E(X^2)= \frac{1}{B(α,β)}∫_{0}^{1}x^{(α+2)-1}(1-x)^{β-1}dx=\frac{1}{B(α,β)}B(α+2,β)$$
+ **Reemplazando se tiene que**
+ $B(α+1,β)=\frac{α}{(α+β)}B(α,β)$

-  $B(α+2,β)=\frac{α+1}{(α+1+β)}B(α+1,β)=\frac{α+1}{(α+1+β)}\frac{α}{(α+β)}B(α,β)$

 $$E(X^2)=\frac{α+1}{(α+1+β)}\frac{α}{(α+β)}$$

+ **Verifique que**
 + $$σ^2 = E(X^2)-E(X)^2=\frac{α+1}{(α+1+β)}\frac{α}{(α+β)}-(\frac{α}{(α+β)})^2$$
 + $$σ^2 =\frac{αβ}{(α+1+β)(α+β)^2}$$

#**a. Construya el gráfico de la Distribución para   $α=50$ y $β=49$**
+ **Función de Densidad**. Si $X$ es una variable aleatoria continua, se dice que $X$ tiene una distribución Beta con parámetros ${\displaystyle \alpha ,β >0}$ y escribimos $X∼Beta(α,β)$ si su función de densidad es

$$f(x;\alpha, \beta)=\frac{1}{B(α,β)}x^{α-1}(1-x)^{β-1}, \ \ 0<x<1$$

- Donde $$
{B(α,β)} = \int_0^{1} x^{α-1}(1-x)^{β-1}dx=\frac{\Gamma(α)\Gamma(β)}{\Gamma(α+β)},.
$$

```python
#**a. Construya el gráfico de la Distribución dada si  $α=50$ y $β=49$**
from numpy.ma.core import log
from sympy import integrate, init_printing
from sympy.abc import x
# nuestra libreria
import numpy as np
import matplotlib.pyplot as plt
# Definimos una funcion
a = 50 # parametro alpha
b = 49 # parametro beta
beta = factorial(a-1)*factorial(b-1)/factorial(a+b-1)

def dis_beta(x):
  return x**(a-1)*(1-x)**(b-1)/(beta)
# los valores que va a tomar x
x = np.linspace(0,1,1000)
fig, ax = plt.subplots() #lugar donde vamos a estar graficando, ax objeto y subplot porque vamos hacer varios graficos
ax.plot(x,dis_beta(x)) # variable independiente y la funcion
plt.title('Fabrica A - Distribución Beta α=50 y β=49')
plt.ylabel('Probabilidad')
plt.xlabel('Produccion perdida (Porcentajes)')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

```python
#**a. Construya el gráfico de la Distribución dada si  α=11.7 y β=7.24**
plt.figure(figsize=(8, 8))
a = 50 # parametro alpha
b = 49 # parametro beta

beta = stats.beta(a, b)  # alfa 50, beta = 49
x = np.linspace(beta.ppf(0.0001),
                beta.ppf(0.99999), 10000)
fp = beta.pdf(x) # Función de Probabilidad
plt.plot(x, fp)

plt.title('Fabrica A - Distribución Beta α=50 y β=49')
plt.ylabel('Probabilidad')
plt.xlabel('Produccion perdida (Porcentajes)')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

```python
#**d. Valores críticos. **
x = np.linspace(0, 1, 401)
y = stats.beta(50, 49).pdf(x)
a = round(stats.beta(50, 49).ppf(0.785),5)
fig, ax = plt.subplots(figsize=[6, 4])
c = 1 - stats.beta(a=50, b=49).cdf(a)
x1 = np.linspace(a, 1, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.beta(50, 49).pdf(x1) , color="r")

ax.text(a, 2, f'Área pedida ={c:.4f}', horizontalalignment='center')

ax.text(a, 0.00001, f'valor pedido es {a}')

plt.title('Fabrica A - Distribución Beta α=50 y β=49')
plt.ylabel('Probabilidad')
plt.xlabel('Produccion perdida (Porcentajes)')
ax.grid() # le da mas elegancia al grafico
plt.show()

print('Produccion perdida es igual a',a,'dias')
print(f'El área a la derecha de P(X>={a})=',f'{c:.4f}')
```

#  **m. Dada la Fábrica A. Parámetros de la Distribución de Beta   $α=11.7$ y $β=10.52$, calcule el valor de $k$  tal que $P (k < X < 0.55) = 0.42$**

```python
# **m**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(0, 1, 401)

y = stats.beta(11.7,10.52).pdf(x)
a = 0.55 # Valor de la variable a calcular a la izquierda
b = stats.beta(11.7,10.52).cdf(a)
c = stats.beta(11.7,10.52).cdf(a) - 0.42
d = round(stats.beta(11.7,10.52).ppf(c),2) # valor critico pedido
e = stats.beta(11.7,10.52).cdf(d) # Probabilidad acumulada
x1 = np.linspace(0, d, 401)
x2 = np.linspace(d, a, 401)
x3 = np.linspace(a, 1, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.beta(11.7,10.52).pdf(x1) , color="r", alpha = 0.3)
ax.fill_between(x2, 0,stats.beta(11.7,10.52).pdf(x2) , color="b", alpha =0.3)
ax.fill_between(x3, 0,stats.beta(11.7,10.52).pdf(x3) , color="y", alpha =0.3)

ax.text(d, 0.00001, d)
ax.text(a, 0.00001, a)

ax.text(a, 1, f'Area1= {1-b:.4f}')
ax.text((a+d)/2, 1.2, f'Area2= {c:.4f}')
ax.text(d, 0.3, f'Area3= {e:.4f}')


print(f'Area solicitada = {c:.4f}')

print(f'El valor critico solicitado es k ={d}')
print(f'P(X<={d})={e:.4f}')
print(f'P({d}=<X<={a})={b-e:.4f}')
print(f'P(X<={a})={b:.4f}')

plt.title('Fabrica A - Distribución beta(11.7,10.52)')
plt.ylabel('Probabilidad')
plt.xlabel('Produccion perdida (Porcentajes)')
ax.grid() # le da mas elegancia al grafico
plt.show()
```
