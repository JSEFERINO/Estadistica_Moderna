```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Datos
Autoestima = [10, 17, 12, 50, 30, 35, 16, 13, 20, 24, 45, 30, 34, 15, 23, 15, 33, 29, 28, 36]
Desempeño = [12, 15, 15, 80, 20, 28, 17, 18, 18, 16, 70, 40, 26, 18, 20, 16, 24, 25, 18, 40]

# Realizar el ajuste de la regresión lineal
slope, intercept, r_value, p_value, std_err = stats.linregress(Autoestima, Desempeño)

# Calcular el coeficiente de determinación (R-cuadrado)
r_squared = r_value**2

# Crear un rango de valores para la línea de regresión
x = np.linspace(min(Autoestima), max(Autoestima), 100)
y = slope * x + intercept

# Crear el gráfico de dispersión
plt.scatter(Autoestima, Desempeño, label='Datos')

# Agregar la línea de regresión
plt.plot(x, y, color='red',
         label=f'Regresión Lineal\n$R^2={r_squared:.3f}$\ny = {slope:.2f} * x + {intercept:.2f}\n r = {r_value:.2f}')

# Etiquetas y leyenda
plt.xlabel('Autoestima')
plt.ylabel('Desempeño')
plt.legend()

# Mostrar el gráfico
plt.show()

# Imprimir los coeficientes
print(f"Coeficiente de correlación: {r_value:.2f}")
print(f"Coeficiente de determinación (R-cuadrado): {r_squared:.2f}")
```
