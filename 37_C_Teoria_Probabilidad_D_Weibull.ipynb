# **a. Generalidades de la Ditribución Weibull**
+ **Wikipedia**. La distribución de Weibull es una distribución de probabilidad continua. Recibe su nombre de Waloddi Weibull, que la describió detalladamente en 1951, aunque fue descubierta inicialmente por Fréchet (1927) y aplicada por primera vez por Rosin y Rammler (1933) para describir la distribución de los tamaños de determinadas partículas.

+ **Ejemplos para la distribución Weibull**: Veamos algunos casos donde esta distribución es aplicable:

 - Análisis de la supervivencia.
 - En ingeniería, para modelar procesos estocásticos relacionados con el tiempo de fabricación y distribución de bienes.

 - Aplicación de la distribución de probabilidad acumulada de Weibull a lluvias diárias máximas.
 - Teoría de valores extremos
 - Meteorología
 - Para modelar la distribución de la velocidad del viento (frecuencia con la que se dan diferentes velocidades de viento)
 - En telecomunicaciones
 - En sistemas de radar para simular la dispersión de la señal recibida
 - En energía solar, para modelar la distribución de irradiación solar anual
 - En seguros, para modelar el tamaño de las pérdidas
 - En la hidrología, se utiliza la distribución de Weibull para analizar variables aleatorias como valores máximos de la precipitación y la descarga de ríos,​ y además para describir épocas de sequía.

# **b. La Distribución Weibull - Definición**

+ **Función de Densidad**. Si $X$ es una variable aleatoria continua, se dice que $X$ tiene una distribución Weibull con parámetros ${\displaystyle \alpha ,β >0}$ y escribimos $X∼Weibull(α,β)$ si su función de densidad es

$${\displaystyle f(x;\alpha, \beta)=\frac{αx^{α-1}}{β^α}e^{-(\frac{x}{β})^{\alpha }}x\geq 0}$$


+ Tambien $$E(X)=\mu=β\Gamma(1+\frac{1}{α}),  \ \ \ \ V(X)=σ^2=β^2[\Gamma(1+\frac{2}{α})-\Gamma^2(1+\frac{1}{α})]$$

+ En dónde $ \{X\leq x\}$ representa al evento de todos resultados menores que $x$.
+ $\mu$ representa al número esperado.
+ $V(X)$ representa a la varianza.
+ Se puede mostrar

+ La *función Gamma* satisface $\Gamma(n) = (n-1)!\quad\forall n\in\mathbb N$ a través de la integral de Euler

$$
\Gamma(α) = \int_0^\infty x^{α-1}e^{-x}dx\,.
$$


# **c. Veamos que es una función de densidad**
+ $$ ∫_{-∞}^{∞}f(x)dx= ∫_{0}^{∞}\frac{α}{β^α}x^{α-1}exp{\{\frac{-x}{β}\}^α}dx$$

+ **hacemos** $y = (\frac{x}{β})^α$ esto es $x = βy^{1/α} ⇒dx= (β/α)y^{1/α-1}dy$
+  $$ ∫_{-∞}^{∞}f(x)dx=\frac{α}{β^α}∫_{0}^{∞}(βy^{1/α})^{α-1}e^{-y}(β/α)y^{1/α-1}dy$$
+ **Simplificando**

+  $$ ∫_{-∞}^{∞}f(x)dx=\frac{1}{1}∫_{0}^{∞}(y^{1/α})^{α-1}e^{-y}y^{1/α-1}dy$$

+  $$ ∫_{-∞}^{∞}f(x)dx=∫_{0}^{∞}(y^{1-1/α})e^{-y}y^{1/α-1}dy=∫_{0}^{∞}e^{-y}dy=1$$

# **d. Valor Esperado**
+ $$E(X)=μ= ∫_{-∞}^{∞}xf(x)dx= ∫_{0}^{∞}x\frac{α}{β^α}x^{α-1}exp{\{\frac{-x}{β}\}^α}dx$$

+ $$E(X)= ∫_{0}^{∞}\frac{α}{β^α}x^{α}exp{\{\frac{-x}{β}\}^α}dx$$

+ **hacemos** $y = (\frac{x}{β})^α$ esto es $x = βy^{1/α} ⇒dx= (β/α)y^{1/α-1}dy$
+ $$E(X)=μ=\frac{α}{β^α}∫_{0}^{∞}(βy^{1/α})^{α}e^{-y}(β/α)y^{1/α-1}dy$$
+ **Simplificando**

+  $$E(X)=\frac{β}{1}∫_{0}^{∞}(y^{1/α})^{α}e^{-y}y^{1/α-1}dy$$

+  $$E(X)=β∫_{0}^{∞}(y^{1})e^{-y}y^{1/α-1}dy=β∫_{0}^{∞}y^{(1/α+1)-1}e^{-y}dy$$

+ **De la definición**
$$
\Gamma(α) = \int_0^\infty x^{α-1}e^{-x}dx\,.
$$


+  $$E(X)=β\Gamma(\frac{1}{α}+1)$$



# **e. Varianza de una Variable aleatoria Weibull**
+ **Usando el mismo procedimiento arriba, pruebe que** $$E(X^2)=μ= ∫_{-∞}^{∞}x^2f(x)dx= ∫_{0}^{∞}x^2\frac{α}{β^α}x^{α-1}exp{\{\frac{-x}{β}\}^α}dx=β^2\Gamma(\frac{2}{α}+1)$$
+ Por tanto $$V(X)=β^2\Gamma(\frac{2}{α}+1)-β^2\Gamma^2(\frac{1}{α}+1)$$

#**a. Construya el gráfico de la Distribución para la Región Montañosa si  $α=11.7$ y $β=7.24$**
+ Una variable aleatoria continua $X$ sigue una distribución *Gamma* de parámetros $β$, lo cual denotamos por $$X∼Weibull(α,β)$$ si tiene funcion FDP dada por la formula:
$${\displaystyle f(x;\alpha, \beta)=\frac{αx^{α-1}}{β^α}e^{-(\frac{x}{β})^{\alpha }}x\geq 0}$$


```python
#**a. Construya el gráfico de la Distribución dada si  $α=11.7$ y $β=7.24$**
from numpy.ma.core import log
from sympy import integrate, init_printing
from sympy.abc import x
# nuestra libreria
import numpy as np
import matplotlib.pyplot as plt
# Definimos una funcion
a = 11.7 # parametro alpha
b = 7.24 # parametro beta

def dis_weibull(x):
  return np.exp(-(x/b)**a)*a*(x**(a-1))/(b**a)
# los valores que va a tomar x
x = np.linspace(0,10,1000)
fig, ax = plt.subplots() #lugar donde vamos a estar graficando, ax objeto y subplot porque vamos hacer varios graficos
ax.plot(x,dis_weibull(x)) # variable independiente y la funcion
plt.title('Región Montañosa - Distribución Weibull α=11.7 y β=7.24')
plt.ylabel('Probabilidad')
plt.xlabel('Carga Acumulada ESAL (millones)')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

```python
#**a. Construya el gráfico de la Distribución dada si  α=11.7 y β=7.24**
plt.figure(figsize=(8, 8))
scale = 15 # parametro
weibull = stats.weibull_min(c = 11.7, scale=7.24)  # alfa 11.7, beta = 7.24
x = np.linspace(weibull.ppf(0.0001),
                weibull.ppf(0.99999), 10000)
fp = weibull.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Región Montañosa - Distribución Weibull c=α=11.7 y β=7.24')
plt.ylabel('Probabilidad')
plt.xlabel('Carga Acumulada ESAL (millones)')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

#**d. Valores críticos. La carga acumulada correspondientes al 21.5% de cola superior para la Región Montañosa es**

```python
#**d. Valores críticos. **
x = np.linspace(0, 10, 401)
y = stats.weibull_min(c = 11.7, scale=7.24).pdf(x)
a = round(stats.weibull_min(c = 11.7, scale=7.24).ppf(0.785),3)
fig, ax = plt.subplots(figsize=[6, 4])
c = 1 - stats.weibull_min(c = 11.7, scale=7.24).cdf(a)
x1 = np.linspace(a, 10, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.weibull_min(c = 11.7, scale=7.24).pdf(x1) , color="r")

ax.text(a, 0.1, f'Área pedida ={c:.4f}', horizontalalignment='center')

ax.text(a, 0.00001, f'valor pedido es {a}')

plt.title('Región Montañosa - Distribución Weibull c=α=11.7 y β=7.24')
plt.ylabel('Probabilidad')
plt.xlabel('Carga Acumulada ESAL (millones)')
ax.grid() # le da mas elegancia al grafico
plt.show()

print('Carga Acumulada ESAL es igual a',a,'dias')
print(f'El área a la derecha de P(X>={a})=',f'{c:.4f}')
```

```python
# **m. Dada una distribución exponencial  con $μ=15$, calcule el valor de k tal que P (k < X < 7.5) = 0.42**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(0, 20, 401)

y = stats.weibull_min(c = 13.88, scale=7.24).pdf(x)
a = 7.5 # Valor de la variable a calcular a la izquierda
b = stats.weibull_min(c = 13.88, scale=7.24).cdf(a)
c = stats.weibull_min(c = 13.88, scale=7.24).cdf(a) - 0.42
d = round(stats.weibull_min(c = 13.88, scale=7.24).ppf(c),2) # valor critico pedido
e = stats.weibull_min(c = 13.88, scale=7.24).cdf(d) # Probabilidad acumulada
x1 = np.linspace(0, d, 401)
x2 = np.linspace(d, a, 401)
x3 = np.linspace(a, 20, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.weibull_min(c = 13.88, scale=7.24).pdf(x1) , color="r", alpha = 0.3)
ax.fill_between(x2, 0,stats.weibull_min(c = 13.88, scale=7.24).pdf(x2) , color="b", alpha =0.3)
ax.fill_between(x3, 0,stats.weibull_min(c = 13.88, scale=7.24).pdf(x3) , color="y", alpha =0.3)

ax.text(d-1, 0.00001, d)
ax.text(a, 0.00001, a)

ax.text(10, 0.1, f'Area1= {1-b:.4f}')
ax.text(d-3, 0.2, f'Area2= {c:.4f}')
ax.text(0.1, 0.3, f'Area3= {e:.4f}')


print(f'Area solicitada = {c:.4f}')

print(f'El valor critico solicitado es k ={d}')
print(f'P(X<={d})={e:.4f}')
print(f'P(k=<X<=130)={b-e:.4f}')
print(f'P(X<=130)={b:.4f}')

plt.title('Todas con el mismo alfa - Distribución Weibull')
plt.ylabel('Probabilidad')
plt.xlabel('Carga Acumulada ESAL (millones)')
ax.grid() # le da mas elegancia al grafico
plt.show()
```
