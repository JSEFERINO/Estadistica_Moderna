# **Paso_3. Variabilidad Total y Varianza Total**


## **a. La variabilidad total es _SST_ y la Varianza Total _CMT_**

```python
## **b. La variabilidad total es _SST_ y la Varianza Total _CMT_**
SST = sum_Y2 - sum_T**2/N
MST = SST/(N-1)
print(f'La variabilidad total es SST = {SST}')
print(f'La Varianza total es MST = {MST}')
```

## **b. La variabilidad de los Tratamientos es _SSTr_ y la Varianza Total _CMTr_**

```python
## **c. La variabilidad de los Tratamientos es _SSTr_ y la Varianza Total _CMTr_
SSA = sum_A22/(n*J) - sum_T**2/N

MSA = SSA/(I-1)
print(f'La variabilidad de los tratamientos es SSA = {SSA}')
print(f'La variabilidad de los tratamientos es MSA = {MSA}')
```

```python
## **c. La variabilidad de los Tratamientos es _SSTr_ y la Varianza Total _CMTr_
SSB = sum_B22/(n*I) - sum_T**2/N

MSB = SSB/(J-1)
print(f'La variabilidad de los tratamientos es SSB = {SSB}')
print(f'La variabilidad de los tratamientos es MSB = {MSB}')
```

```python
## **c. La variabilidad de los Tratamientos es _SSTr_ y la Varianza Total _CMTr_
SSAB = sum_AB2/n - sum_T**2/N-SSA-SSB

MSAB = SSAB/((I-1)*(J-1))
print(f'La variabilidad de los tratamientos es SSAB = {SSAB}')
print(f'La variabilidad de los tratamientos es MSAB = {MSAB}')
```

## **c. La variabilidad de los Residuos es _SSE_ y la Varianza Total _CME_**

```python
SSE = SST - SSA - SSB - SSAB

MSE = SSE/(I*J*(n-1))
print(f'La variabilidad de los residuos es SSE = {SSE}')
print(f'La varianza de los residuos es MSE = {MSE}')
```

## **d. El coeficiente de determinación del ajuste**

$$R^2 = 1- \frac{SSE}{SST}$$

```python
R_2 = 1- SSE/SST
print(f'El coeficente de R_cuadrado es {R_2*100} por ciento')
```

## **f. El valor p, de la prueba**

```python
from scipy import stats
F1 = stats.f(I-1,I*J*(n-1))
F2 = stats.f(J-1,I*J*(n-1))
F3 = stats.f((I-1)*(J-1),I*J*(n-1))
valor_p1 = 1 - F1.cdf(MSA/MSE)
valor_p2 = 1 - F2.cdf(MSB/MSE)
valor_p3 = 1 - F3.cdf(MSAB/MSE)
```

## **g. Construyendo la Tabla Anova**

```python
from texttable import Texttable
t3 = Texttable()
t3.add_rows([['Fuente','SS', 'g.l', 'MS', 'F', 'valor_p'],
           ['A',	SSA,	I-1,	MSA,	MSA/MSE, valor_p1],
           ['B',	SSB,	J-1,	MSB,	MSB/MSE, valor_p2],
           ['AB',	SSAB,	(I-1)*(J-1),	MSAB,	MSAB/MSE, valor_p3],
           ['Error',	SSE,	I*J*(n-1),	MSE, '', ''],
           ['Total',	SST,	N-1, '' , '', '']])
print(t3.draw())
```
