$$Y = mX+b$$

## **c. Estadisticos Usados en la regresión**
+ $$suma(x)=\sum_{i=1}^{n}x_i=$$

*  $$suma(y)=\sum_{i=1}^{n}y_i=$$
*  $$suma(xy)=\sum_{i=1}^{n}x_iy_i=$$

*   $$suma(x^2)=\sum_{i=1}^{n}x^{2}_i=$$
*   $$suma(y^2)=\sum_{i=1}^{n}y^{2}_i=$$


*   $$s_{xy} = \sum_{i=1}^{n}(x_i-\bar{x})(y_i-\bar{y})=\sum_{i=1}^{n}x_iy_i-n\bar{x}\bar{y}=$$
*   $$s_x = \sqrt{\sum_{i=1}^{n}(x_i-\bar{x})^2}= \sqrt{\sum_{i=1}^{n}(x_i^2-n\bar{x}^2)}=$$
*   $$s_y = \sqrt{\sum_{i=1}^{n}(y_i-\bar{y})^2}= \sqrt{\sum_{i=1}^{n}(y_i^2-n\bar{y}^2)}=$$


*   Coeficiente de correlación $$r = \frac{s_{xy}}{s_xs_y}$$
*   Coeficientes de regresión $$b_1 = \frac{s_{xy}}{s^2_{xx}}$$


*   Coeficientes de regresión $$b_0 = \bar{y}-b_1\bar{x}$$
