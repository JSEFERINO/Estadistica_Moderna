# **b. La función Gamma**
+ La *función Gamma se define*  

$$
\Gamma(α) = \int_0^\infty x^{α-1}e^{-x}dx\,.
$$
# **c. Propiedades de la función Gamma**
+ **Propiedad 1**. $\Gamma(1) =\Gamma(2) =1$
 - **Prueba**
 $$
\Gamma(1) = ∫_{0}^{∞} x^{1-1}e^{-x}dx= ∫_{0}^{∞} e^{-x}dx=-e^{-x}|_{0}^{∞}=1
$$
$$
\Gamma(2) = ∫_{0}^{∞} x^{2-1}e^{-x}dx= ∫_{0}^{∞} xe^{-x}dx=-xe^{-x}|_{0}^{∞}+∫_{0}^{∞} e^{-x}dx=1
$$
+ **Propiedad 2**. $\Gamma(α+1) = α!$, si $α\in\mathbb N$
+ **Prueba**
  $$
\Gamma(α+1) = ∫_{0}^{∞} x^{α-1+1}e^{-x}dx= ∫_{0}^{∞} x^{α}e^{-x}dx=
 \ \ \ -x^{α}e^{-x}|_{0}^{∞}+α∫_{0}^{∞} x^{α-1}e^{-x}dx=α\Gamma(α)
$$
+   $$
\Gamma(α+1) =α\Gamma(α)=α(α-1)\Gamma(α-1)=α(α-1)⋯2\cdot\Gamma(1)=α!,α\in\mathbb N
$$

+ **Propiedad 3**. $\Gamma(1/2) = \sqrt{π}$
$$
\Gamma(1/2) = ∫_{0}^{∞} x^{1/2-1}e^{-x}dx= ∫_{0}^{∞}  x^{-1/2}e^{-x}dx
$$
+ **Hagamos** $u^2=2x$ entonces $2udu=2dx$ y $x=\frac{u^2}{2}$ y $x^{-1/2}=(\frac{u^2}{2})^{-1/2}=\frac{\sqrt{2}}{u}$
+ **Reemplazando todo en la integral arriba**

$$
\Gamma(1/2) = ∫_{0}^{∞} x^{1/2-1}e^{-x}dx= ∫_{0}^{∞}  x^{-1/2}e^{-x}dx=∫_{0}^{∞}\frac{\sqrt{2}}{u}e^{-\frac{u^2}{2}}udu=∫_{0}^{∞}\sqrt{2}e^{-\frac{u^2}{2}}du
$$
$$
\Gamma(1/2) = ∫_{0}^{∞}\sqrt{2}e^{-\frac{u^2}{2}}du=\frac{1}{2}∫_{-∞}^{∞}\sqrt{2}e^{-\frac{u^2}{2}}du=∫_{-∞}^{∞}\frac{1}{\sqrt{2}}e^{-\frac{u^2}{2}}du =\sqrt{π}∫_{-∞}^{∞}\frac{1}{\sqrt{2π}}e^{-\frac{u^2}{2}}du=\sqrt{π}*1=\sqrt{π}
$$

# **180. Distribuciones continuas - La Distribución Gamma**
# **a. La Ditribución Gamma - Definición**

+ Una variable aleatoria continua $X$ sigue una distribución *Gamma* de parámetros $β$, lo cual denotamos por $$X∼gamma(α,β)$$ si tiene funcion FDP dada por la formula:
$$ f(x,α,β) = \frac{1}{β^αΓ(α)}x^{α-1}exp{\{\frac{-x}{β}\}}, \ x\geq 0, 0<α,β<∞, $$
+ Si hacemos $λ=β^{-1}$ Se tendría que:

$$ f(x,α,λ) = \frac{(λx)^{α-1}}{Γ(α)}λe^{-λx}, \ x\geq 0$$

+ Tambien $$E(X)=\mu=βα,  \ \ \ \ V(X)=σ^2=αβ^2$$

+ En dónde $ \{X\leq x\}$ representa al evento de todos resultados menores que $x$.
+ $\mu$ representa al número esperado.
+ $V(X)$ representa a la varianza.
+ Se puede mostrar



# **b. Veamos que es una función de densidad**
+ $$ ∫_{-∞}^{∞}f(x)dx= ∫_{0}^{∞}\frac{1}{β^αΓ(α)}x^{α-1}exp{\{\frac{-x}{β}\}}dx= \frac{1}{β^αΓ(α)}∫_{0}^{∞}x^{α-1}exp{\{\frac{-x}{β}\}}dx$$

+ **hacemos** $y = \frac{x}{β}$ esto es $x = βy ⇒dx= βdy$
+  $$ ∫_{-∞}^{∞}f(x)dx=\frac{1}{β^αΓ(α)}∫_{0}^{∞}(βy)^{α-1}e^{-y}βdy
=\frac{β^α}{β^αΓ(α)}∫_{0}^{∞}y^{α-1}e^{-y}dy
$$
+ **Pero sabemos que**
$$
\Gamma(α) = \int_0^\infty x^{α-1}e^{-x}dx\,.
$$
+ **De esta forma se tiene**
 $$ ∫_{-∞}^{∞}f(x)dx=\frac{1}{β^αΓ(α)}∫_{0}^{∞}(βy)^{α-1}e^{-y}βdy
=\frac{β^α}{β^αΓ(α)}\Gamma(α) =1
$$


# **c. Valor Esperado**
+ $$E(X)=μ= ∫_{-∞}^{∞}xf(x)dx= ∫_{0}^{∞}\frac{x}{β^αΓ(α)}x^{α-1}exp{\{\frac{-x}{β}\}}dx$$

+ $$E(X)= ∫_{0}^{∞}\frac{αβ}{β^{α+1}αΓ(α)}x^{(α+1)-1}exp{\{\frac{-x}{β}\}}dx
\ =αβ ∫_{0}^{∞}\frac{1}{β^{α+1}Γ(α+1)}x^{(α+1)-1}exp{\{\frac{-x}{β}\}}dx
$$

+ **Por la parte arriba se tiene**
$$ ∫_{-∞}^{∞}f(x)dx= ∫_{0}^{∞}\frac{1}{β^{α+1}Γ(α+1)}x^{(α+1)-1}exp{\{\frac{-x}{β}\}}dx=1$$

+ **De manera que**

+ $$E(X)=αβ $$


# **d. Varianza de una Variable aleatoria normal**

+  $$V(X)=σ^2$$
 - $$E(X^2)= ∫_{-∞}^{∞}x^2f(x)dx= ∫_{0}^{∞}\frac{x^2}{β^αΓ(α)}x^{α-1}exp{\{\frac{-x}{β}\}}dx$$

+ $$E(X^2)= ∫_{0}^{∞}\frac{α(α+1)β^2}{β^{α+2}α(α+1)Γ(α)}x^{(α+2)-1}exp{\{\frac{-x}{β}\}}dx
\ =α(α+1)β^2 ∫_{0}^{∞}\frac{1}{β^{α+2}Γ(α+2)}x^{(α+2)-1}exp{\{\frac{-x}{β}\}}dx
$$

+ **Por la parte arriba se tiene**
$$ ∫_{-∞}^{∞}f(x)dx= ∫_{0}^{∞}\frac{1}{β^{α+2}Γ(α+2)}x^{(α+2)-1}exp{\{\frac{-x}{β}\}}dx=1$$

+ **De manera que**

+ $$E(X^2)=α(α+1)β^2 $$

+ **Aplicando la formula de varianza**
 +  $$V(X)=σ^2 = E(X^2)-E(X)^2=α(α+1)β^2-α^2β^2=αβ^2$$

#**a. Construya el gráfico de la Distribución dada si  $α=5$ y $β=8$**
+ Una variable aleatoria continua $X$ sigue una distribución *Gamma* de parámetros $β$, lo cual denotamos por $$X∼gamma(α,β)$$ si tiene funcion FDP dada por la formula:
$$ f(x,α,β) = \frac{1}{β^αΓ(α)}x^{α-1}exp{\{\frac{-x}{β}\}}, \ x\geq 0, 0<α,β<∞, $$


```python
#**a. Construya el gráfico de la Distribución dada si  $α=5$ y $β=8$**
from numpy.ma.core import log
from sympy import integrate, init_printing
from sympy.abc import x
# nuestra libreria
import numpy as np
import matplotlib.pyplot as plt
# Definimos una funcion
a = 5 # parametro alpha
b = 8 # parametro beta

def dis_gamma(x):
  return np.exp(-x/b)*(x**(a-1))/((b**a)*factorial(a-1))
# los valores que va a tomar x
x = np.linspace(0,200,1000)
fig, ax = plt.subplots() #lugar donde vamos a estar graficando, ax objeto y subplot porque vamos hacer varios graficos
ax.plot(x,dis_gamma(x)) # variable independiente y la funcion
plt.title('Distribución Gamma α=5 y β=8')
plt.ylabel('Probabilidad')
plt.xlabel('Consumo de Energia Kw/hora')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

```python
#**a. Construya el gráfico de la Distribución dada si  α=5 y β=8**
plt.figure(figsize=(8, 8))
scale = 15 # parametro
gamma = stats.gamma(a=5, scale=8)  # alfa 3, beta = 8
x = np.linspace(gamma.ppf(0.0001),
                gamma.ppf(0.99999), 10000)
fp = gamma.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Gamma α=5 y β=8')
plt.ylabel('Probabilidad')
plt.xlabel('Consumo de Energia Kw/hora')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

```python
# **m. Dada una distribución exponencial  con $μ=15$, calcule el valor de k tal que P (k < X < 70) = 0.42**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(0, 250, 401)

y = stats.gamma(a=10, scale=10).pdf(x)
a = 130 # Valor de la variable a calcular a la izquierda
b = stats.gamma(a=10, scale=10).cdf(a)
c = stats.gamma(a=10, scale=10).cdf(a) - 0.42
d = round(stats.gamma(a=10, scale=10).ppf(c),2) # valor critico pedido
e = stats.gamma(a=10, scale=10).cdf(d) # Probabilidad acumulada
x1 = np.linspace(0, d, 401)
x2 = np.linspace(d, a, 401)
x3 = np.linspace(a, 250, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.gamma(a=10, scale=10).pdf(x1) , color="r", alpha = 0.3)
ax.fill_between(x2, 0,stats.gamma(a=10, scale=10).pdf(x2) , color="b", alpha =0.3)
ax.fill_between(x3, 0,stats.gamma(a=10, scale=10).pdf(x3) , color="y", alpha =0.3)

ax.text(d, 0.00001, d)
ax.text(a, 0.00001, a)

ax.text(40, 0.008, f'Area1= {1-b:.4f}')
ax.text(d+5, 0.001, f'Area2= {c:.4f}')
ax.text(0.1, 0.005, f'Area3= {e:.4f}')
plt.title('Distribución gamma β=10')
plt.ylabel('Probabilidad')
plt.xlabel('Consumo de Energia Kw/hora')
ax.grid() # le da mas elegancia al grafico
print(f'Area solicitada = {c:.4f}')

print(f'El valor critico solicitado es k ={d}')
print(f'P(X<={d})={e:.4f}')
print(f'P(k=<X<=130)={b-e:.4f}')
print(f'P(X<=130)={b:.4f}')

plt.show()
```
