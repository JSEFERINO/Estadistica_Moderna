# **b. Definición - Distribución Chi-Cuadrada.**
### **Como la suma de normales estándar**
- Sean ${\displaystyle Z_{1},\dots ,Z_{k}}$ variables aleatorias independientes tales que ${\displaystyle Z_{i}\sim N(0,1)}$ para ${\displaystyle i=1,2,\dots ,k}$ entonces la variable aleatoria $X$ definida por

$${\displaystyle {\begin{aligned}X&=Z_{1}^{2}+Z_{2}^{2}+\cdots +Z_{k}^{2}\\&=\sum _{i=1}^{k}Z_{i}^{2}\end{aligned}}}$$
- tiene una distribución chi cuadrada con $k$ grados de libertad.

### **Notación**.
+  Si la variable aleatoria continua $X$ tiene una distribución Chi Cuadrada con $k$ grados de libertad entonces escribiremos ${\displaystyle X\sim \chi _{k}^{2}}$.

###  **Función de Densidad**
+ Si $X$ es una variable aleatoria continua, se dice que $X$ tiene una distribución Chi-cuadrado con parámetro $n$ y escribimos $X∼Χ^2(n)$ si su función de densidad es

$${\displaystyle f(x;n)=\frac{{(\frac{x}{2})^{\frac{n}{2}-1}}e^{-(\frac{x}{2})}}{2\Gamma(\frac{n}{2})},\dots ,x\geq 0}$$


+ Tambien $$E(X)=\mu=n,  \ \ \ \ V(X)=σ^2=2n$$





# **c. Distribuciones continuas - La Distribución Chi-Cuadrado**



## **Variable aleatoria con Distribución $χ^2$**

- Sea ${\displaystyle X_{1},\dots ,X_{n}}$ una muestra aleatoria proveniente de una población con distribución ${\displaystyle N(\mu ,\sigma ^{2})}$ entonces

  - ${\displaystyle {\overline {X}}}$ y el vector ${\displaystyle \left(X_{1}-{\overline {X}},\dots ,X_{n}-{\overline {X}}\right)}$ son independientes.
  - ${\displaystyle {\overline {X}}}$ y ${\displaystyle S^{2}}$ son independientes.
  - Entonces, la variable aleatoria $${\displaystyle {\frac {(n-1)S^{2}}{\sigma ^{2}}}\sim \chi _{n-1}^{2}}$$
  - ${\displaystyle \operatorname {E} [S^{2}]=\sigma ^{2}}$ y ${\displaystyle \operatorname {Var} (S^{2})={\frac {2\sigma ^{4}}{n-1}}}$.
- donde ${\displaystyle {\overline {X}}={\frac {1}{n}}\sum _{i=1}^{n}X_{i}}$ y ${\displaystyle S^{2}={\frac {1}{n-1}}\sum _{i=1}^{n}\left(X_{i}-{\overline {X}}\right)^{2}}$ son la media y varianza de la muestra aleatoria respectivamente


## **Intervalo para la varianza con Distribución $χ^2$**
- Sean ${\displaystyle X_{1},\dots ,X_{n}}$ una muestra aleatoria proveniente de una población con distribución ${\displaystyle N(\mu ,\sigma ^{2})}$ donde ${\displaystyle \mu }$ y $\sigma^2$ son desconocidos. Se tiene que ${\displaystyle {\frac {(n-1)S^{2}}{\sigma ^{2}}}\sim \chi _{n-1}^{2}}$
- Sean ${\displaystyle \chi _{n-1,\alpha /2}, \ \chi _{n-1,1-\alpha /2}\in \mathbb {R} }$ tales que ${\displaystyle \operatorname {P} [\chi _{n-1,\alpha /2}<Y<\chi _{n-1,1-\alpha /2}]=1-\alpha }$. siendo ${\displaystyle Y\sim \chi _{n-1}^{2}}$.
$${\displaystyle {\begin{aligned}&\operatorname {P} \left[\chi _{n-1,\alpha /2}<{\frac {(n-1)S^{2}}{\sigma ^{2}}}<\chi _{n-1,1-\alpha /2}\right]=1-\alpha \\&\operatorname {P} \left[{\frac {1}{\chi _{n-1,\alpha /2}}}>{\frac {\sigma ^{2}}{(n-1)S^{2}}}>{\frac {1}{\chi _{n-1,1-\alpha /2}}}\right]=1-\alpha \\&\operatorname {P} \left[{\frac {(n-1)S^{2}}{\chi _{n-1,1-\alpha /2}}}<\sigma ^{2}<{\frac {(n-1)S^{2}}{\chi _{n-1,\alpha /2}}}\right]=1-\alpha \end{aligned}}}$$

- por lo tanto un intervalo de ${\displaystyle (1-\alpha )100\%}$ de confianza para $\sigma ^{2}$ está dado por

$${\displaystyle \left({\frac {(n-1)S^{2}}{\chi _{n-1,1-\alpha /2}}},{\frac {(n-1)S^{2}}{\chi _{n-1,\alpha /2}}}\right)}$$





# **d. Veamos que es una función de densidad**

+ **Densidad Distribución Chi Cuadrado**
$${\displaystyle f(x;n)=\frac{{(\frac{x}{2})^{\frac{n}{2}-1}}e^{-(\frac{x}{2})}}{2\Gamma(\frac{n}{2})},\dots ,x\geq 0}$$

+ **Densidad Distribución Función Gamma**
$$ f(x,α,β) = \frac{1}{β^αΓ(α)}x^{α-1}exp{\{\frac{-x}{β}\}}, \ x\geq 0, 0<α,β<∞, $$


+ **Hagamos las comparaciones en Gamma**
 - $α = \frac{n}{2}$, y $β = 2$
 - $$ f(x,\frac{n}{2},2) = \frac{1}{2^{\frac{n}{2}}Γ(\frac{n}{2})}x^{\frac{n}{2}-1}exp{\{\frac{-x}{2}\}}$$
 - $$ f(x,\frac{n}{2},2) = \frac{1}{2*2^{\frac{n}{2}-1}Γ(\frac{n}{2})}x^{\frac{n}{2}-1}exp{\{\frac{-x}{2}\}}=f(x;n)$$



+ **Conclusión:** $χ^2$ es un caso especial de la función Gamma, reemplazando $α = \frac{n}{2}$, y $β = 2$


# **e. Valor Esperado y varianza**
+ **Valor Esperado de la Gamma es**
 $$E(X)=\mu=βα,  \ \ \ \ V(X)=σ^2=αβ^2$$
+ **Valor Esperado de  $χ^2$**  reemplazando $α = \frac{n}{2}$, y $β = 2$
 $$E(χ^2)=n$$

+ **Varianza de  $χ^2$**  reemplazando $α = \frac{n}{2}$, y $β = 2$
 $$V(χ^2)=2n$$

#**a. Construya el gráfico de la Distribución dada**
+ **Densidad Distribución Chi Cuadrado**
$${\displaystyle f(x;n)=\frac{{(\frac{x}{2})^{\frac{n}{2}-1}}e^{-(\frac{x}{2})}}{2\Gamma(\frac{n}{2})},\dots ,x\geq 0}$$

```python
#**a. Construya el gráfico de la Distribución dada si  $α=50$ y $β=49$**
from numpy.ma.core import log
from sympy import integrate, init_printing
from sympy.abc import x
# nuestra libreria
import numpy as np
import matplotlib.pyplot as plt
# Definimos una funcion
n = 50 # parametro alpha
a = n/2

gamma = factorial(a-1)

def dis_Chi(x):
  return np.exp(-x/2)*(x/2)**(a-1)/(2*gamma)
# los valores que va a tomar x
x = np.linspace(0,100,1000)
fig, ax = plt.subplots() #lugar donde vamos a estar graficando, ax objeto y subplot porque vamos hacer varios graficos
ax.plot(x,dis_Chi(x)) # variable independiente y la funcion

plt.title('Fabrica A - Distribución Chi_Cuadrado')
plt.ylabel('Probabilidad')
plt.xlabel('Variabilidad contenido en contenido de las bolsas')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

```python
#**a. Construya el gráfico de la Distribución dada si  α=11.7 y β=7.24**
plt.figure(figsize=(8, 8))
df = 50 # parametro de forma.
chi2 = stats.chi2(df)

x = np.linspace(chi2.ppf(0.0001),
                chi2.ppf(0.99999), 10000)
fp = chi2.pdf(x) # Función de Probabilidad
plt.plot(x, fp)

plt.title('Fabrica A - Distribución Chi_Cuadrado')
plt.ylabel('Probabilidad')
plt.xlabel('Variabilidad contenido en contenido de las bolsas')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

# **i. Para Fábrica B, calcule la probabilidad de que la variabilidad sea mayor a 60% o menor a 30%**

```python
# **i**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(0, 100, 401)

y = stats.chi2(45).pdf(x)

a = 30 # Valor de la variable a calcular a la izquierda
b = 60 # Valor de la variable a calcular a la izquierda
c = stats.chi2(45).cdf(a) # Probabilidad pedida
d = 1-stats.chi2(45).cdf(b)# Probabilidad pedida
e = c+d # Probabilidad pedida
x1 = np.linspace(0, a, 401)
x2 = np.linspace(b, 100, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.chi2(45).pdf(x1) , color="r")
ax.fill_between(x2, 0,stats.chi2(45).pdf(x2) , color="b")


ax.text(a, 0.00001, a)
ax.text(b, 0.00001, b)

ax.text(a, 0.005, f'Prob1 = {c:.4f}')
ax.text(b, 0.01, f'Prob2 = {d:.4f}')

ax.text((a+b)/2, 0.02, f'Area solicitada = {e:.4f}')
print('la probabilidad pedida es')
print(f'P(X>={b}) +P(X<={a})=',f'Area solicitada = {e:.4f}')


plt.title('Fabrica B - Distribución Chi_Cuadrado')
plt.ylabel('Probabilidad')
plt.xlabel('Variabilidad contenido en contenido de las bolsas')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

# **j. Dada la Fábrica A, calcule el valor de k tal que $P(X>k) = 0.301$**

#  **k. Dada la Fábrica A, calcule el valor de $k$  tal que $P (k < X < 0.55) = 0.42$**

```python
# **k**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(0, 100, 401)

y = stats.chi2(35).pdf(x)
a = 55 # Valor de la variable a calcular a la izquierda
b = stats.chi2(35).cdf(a)
c = stats.chi2(35).cdf(a) - 0.42
d = round(stats.chi2(35).ppf(c),2) # valor critico pedido
e = stats.chi2(35).cdf(d) # Probabilidad acumulada
x1 = np.linspace(0, d, 401)
x2 = np.linspace(d, a, 401)
x3 = np.linspace(a, 100, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.chi2(35).pdf(x1) , color="r", alpha = 0.3)
ax.fill_between(x2, 0,stats.chi2(35).pdf(x2) , color="b", alpha =0.3)
ax.fill_between(x3, 0,stats.chi2(35).pdf(x3) , color="y", alpha =0.3)

ax.text(d, 0.00001, d)
ax.text(a, 0.00001, a)

ax.text(a, 0.01, f'Area1= {1-b:.4f}')
ax.text((a+d)/2, 0.02, f'Area2= {c:.4f}')
ax.text(d, 0.03, f'Area3= {e:.4f}')


print(f'Area solicitada = {c:.4f}')

print(f'El valor critico solicitado es k ={d}')
print(f'P(X<={d})={e:.4f}')
print(f'P({d}=<X<={a})={b-e:.4f}')
print(f'P(X<={a})={b:.4f}')


plt.title('Fabrica A - Distribución Chi_Cuadrado')
plt.ylabel('Probabilidad')
plt.xlabel('Variabilidad contenido en contenido de las bolsas')
ax.grid() # le da mas elegancia al grafico
plt.show()
```
