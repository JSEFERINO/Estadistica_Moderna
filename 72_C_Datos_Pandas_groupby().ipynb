# **Pandas groupby() explicado con ejemplos**
+ De manera similar a la cláusula SQL GROUP BY, DataFrame.groupby() la función pandas se utiliza para recopilar datos idénticos en grupos y realizar funciones agregadas en los datos agrupados. La operación agrupar por implica dividir los datos, aplicar algunas funciones y finalmente agregar los resultados.


+ En pandas , puedes usarlo groupby()con la combinación de sum(), pivot(), transform(), aggregate() y muchos más métodos . En este artículo, cubriré cómo agrupar por una sola columna o por varias columnas mediante el uso de agregaciones con ejemplos.pivot()transform()

# **1. Sintaxis de Pandas groupby()**
+ A continuación se muestra la sintaxis de la groupby()función, esta función toma varios parámetros que se explican a continuación y devuelve DataFrameGroupByun objeto que contiene información sobre los grupos.
+



## **Ejemplo_1. Para explicar varios ejemplos de cómo realizar agrupar por, primero, creemos un DataFrame simple con la combinación de cadenas y columnas numéricas.**

```python
import pandas as pd
technologies   = ({
    'Courses':["Spark","PySpark","Hadoop","Python","Pandas","Hadoop","Spark","Python","NA"],
    'Fee' :[22000,25000,23000,24000,26000,25000,25000,22000,1500],
    'Duration':['30days','50days','55days','40days','60days','35days','30days','50days','40days'],
    'Discount':[1000,2300,1000,1200,2500,None,1400,1600,0]
          })
df = pd.DataFrame(technologies)
print(df)
```

# **2. Ejemplo de pandas groupby()**
+ Como dije anteriormente, _groupby()_la función devuelve DataFrameGroupBy el objeto después de recopilar datos idénticos en grupos de pandas DataFrame.
+ Este objeto contiene varios métodos ( sum(), mean()etc.) que se pueden utilizar para agregar las filas agrupadas.

```python
# Use groupby() to compute the sum
df2 =df.groupby(['Courses']).sum()
print(df2)
```

# **3. Pandas groupby() en dos o más columnas**
+ La mayoría de las veces necesitaríamos realizar la agrupación en varias columnas de DataFrame; puede hacerlo pasando una lista de etiquetas de columnas en las que desea realizar la agrupación.

```python
# Group by multiple columns
df2 =df.groupby(['Courses', 'Duration']).sum()
print(df2)
```

# **4. Agregue índice a los datos agrupados.**
+ De forma predeterminada, _groupby()_ el resultado no incluye el índice de fila; puede agregar el índice utilizando el método DataFrame.reset_index().



```python
# Add Row Index to the group by result
df2 = df.groupby(['Courses','Duration']).sum().reset_index()
print(df2)
```

## **5. Elimine NA /Ninguno/Nan (en la clave de grupo) del resultado**
+ También puede elegir si desea incluir NA/Ninguno/Nan en las claves de grupo o no configurando  dropna el parámetro.
+ De forma predeterminada, el valor de dropna establecido en True. para no incluir valores Ninguno/Nan en el dropna=False parámetro establecido de claves de grupo.

```python
# Drop rows that have None/Nan on group keys
df2=df.groupby(by=['Courses'], dropna=True).sum()
print(df2)
```

# **6. Ordenar el resultado de groupby() por clave de grupo**
+ De forma predeterminada, la función groupby() ordena los resultados por clave de grupo, por lo que llevará más tiempo. Si tiene un problema de rendimiento y no desea ordenar el grupo por el resultado, puede desactivarlo usando el parámetro sort=False.

```python
# Remove sorting on grouped results
df2=df.groupby(by=['Courses'], sort=False).sum()
print(df2)
```

+ **Si desea ordenar las claves en orden descendente, utilice a continuación.**

```python
# Sorting group keys on descending order
groupedDF = df.groupby('Courses',sort=False).sum()
sortedDF=groupedDF.sort_values('Courses', ascending=False)
print(sortedDF)
```

+ **En caso de que desee ordenar por una clave diferente, puede hacerlo utilizando la función DataFrame.apply() .**

```python

# Using apply() & lambda
df.groupby('Courses').apply(lambda x: x.sort_values('Fee'))
```

# **7. Aplicar más agregaciones**

```python
# Groupby & multiple aggregations
result = df.groupby('Courses')['Fee'].aggregate(['min','max', 'mean'])
print(result)
```

```python
# Groupby multiple columns & multiple aggregations
result = df.groupby('Courses').aggregate({'Duration':'count','Fee':['min','max']})
print(result)
```

# **8. Ejemplo completo de groupby() en DataFrame**

```python
# **i. Reescribiendo el DataFrame**
Proveedor = (['A']*4+['B']*4+['C']*4)
Pureza = [94, 95, 93, 96, 88, 92, 86, 87, 95, 96, 96, 92]
df3  = pd.DataFrame({'Proveedor': Proveedor, 'Pureza': Pureza})
df3
```

```python
# Use groupby() to compute the sum
df4 =df3.groupby(['Proveedor']).sum()
print(df4)
```

```python
# Groupby & multiple aggregations
result = df3.groupby('Proveedor')['Pureza'].aggregate(['min','max', 'mean','sum', 'std'])
print(result)
```
