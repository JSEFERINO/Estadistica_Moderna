# **c. La Distribución de Pareto - Definición**

+ Una variable aleatoria continua $X$ sigue una distribución *de Pareto* de parámetros $α$ y $m$, lo cual denotamos por $$X∼Pareto{(α,m)}$$ si tiene funcion FDP dada por la formula:
$$ f(x) = \frac{αm^α}{x^{α+1}}, \ x\geq m, α>0$$


+ Tambien $$E(X)=\mu=\frac{αm}{{α-1}}  \ \ \ \ V(X)=σ^2=\frac{αm^2}{(α-1)^2(α-2)}$$



# **d. Veamos que es una función de densidad**

+ **Densidad Distribución**
$$ f(x) = \frac{αm^α}{x^{α+1}}, \ x\geq m, α>0$$

+ **Integrando**.
$$ ∫_{-∞}^{∞}f(x)dx= ∫_{m}^{∞}\frac{αm^α}{x^{α+1}}dx$$


$$ ∫_{-∞}^{∞}f(x)dx= -\frac{αm^α}{x^{α}α}|_{m}^{∞}=\frac{αm^α}{m^{α}α}=1$$

***
  


# **e. Valor Esperado**

+ **Densidad Distribución**
$$ f(x) = \frac{αm^α}{x^{α+1}}, \ x\geq m, α>0$$

+ **Integrando**.
$$ E(X)=∫_{-∞}^{∞}xf(x)dx= ∫_{m}^{∞}\frac{αm^α}{x^{α}}dx= -\frac{αm^α}{x^{α-1}(α-1)}|_{m}^{∞}= \frac{αm}{(α-1)}$$


***

# **f. Varianza**

+ **Densidad Distribución**
$$ f(x) = \frac{αm^α}{x^{α+1}}, \ x\geq m, α>0$$

+ **Integrando**.
$$ E(X^2)=∫_{-∞}^{∞}x^2f(x)dx= ∫_{m}^{∞}\frac{αm^α}{x^{α-1}}dx= -\frac{αm^α}{x^{α-2}(α-2)}|_{m}^{∞}= \frac{αm^2}{(α-2)}$$

+ **Varianza**

$$ V(X)=E(X^2)- (E(X))^2 = \frac{αm^2}{(α-2)}- (\frac{αm}{(α-1)})^2=\frac{αm^2}{(α-1)^2(α-2)}$$

***

## **Problema.  El ingreso mínimo legal de un sector de la poblacion colombiana es de 500 dolares aproximadamente. Suponga que el ingreso mínimo sigue una distribución de Pareto con un parámetro $α=10$**


#**a. Construya el gráfico de la Distribución dada**

+ **Densidad Distribución**
$$ f(x) = \frac{αm^α}{x^{α+1}}, \ x\geq m, α>0$$

$$f_X(x)=\frac{q_0}{x^{α+1}}$$

```python

#**a.**

# Definimos una funcion
m = 500 # parametro alpha
a = 10

q_0 = a*(m**a)

def dis_Pareto(x):
  return q_0/(x**(a+1))
# los valores que va a tomar x
x = np.linspace(500,1500,1000)
fig, ax = plt.subplots() #lugar donde vamos a estar graficando, ax objeto y subplot porque vamos hacer varios graficos
ax.plot(x,dis_Pareto(x)) # variable independiente y la funcion

plt.title('Distribución Pareto')
plt.ylabel('Probabilidad')
plt.xlabel('X :Salario de un sector')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

```python
#**a. Construya el gráfico de la Distribución dada**
plt.figure(figsize=(8, 8))
m = 500 # parametro alpha
a = 10

P = stats.pareto(10, loc=0, scale=500)

x = np.linspace(P.ppf(0.0001),
                P.ppf(0.99999), 100)
fp = P.pdf(x) # Función de Probabilidad
plt.plot(x, fp)


plt.title('Distribución Pareto')
plt.ylabel('Probabilidad')
plt.xlabel('X :Salario de un sector')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

#   **l. Para el Sector A, calcule el valor de $k$  tal que $P (k < X < 550) = 0.42$**

```python
# **k**

fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(321, 1200, 401)
P = stats.pareto(5, loc=0, scale=321)
y = stats.pareto(5, loc=0, scale=321).pdf(x)

a = 550 # Valor de la variable a calcular a la izquierda
b = P.cdf(a)
c = P.cdf(a) - 0.42
d = round(P.ppf(c),2) # valor critico pedido
e = P.cdf(d) # Probabilidad acumulada
x1 = np.linspace(321, d, 401)
x2 = np.linspace(d, a, 401)
x3 = np.linspace(a, 1200, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,P.pdf(x1) , color="r", alpha = 0.3)
ax.fill_between(x2, 0,P.pdf(x2) , color="b", alpha =0.3)
ax.fill_between(x3, 0,P.pdf(x3) , color="y", alpha =0.3)

ax.text(d, 0.00001, d)
ax.text(a, 0.00001, a)

ax.text(a, 0.002, f'P(X>{a}) = {1-b:.4f}')
ax.text((a+d)/2, 0.008, f'P({d}=<X<={a})= {b-e:.4f}')
ax.text(d, 0.005, f'P(X<={d}) = {e:.4f}')



plt.title('Distribución Pareto')
plt.ylabel('Probabilidad')
plt.xlabel('X :Salario de un sector')
ax.grid() # le da mas elegancia al grafico
plt.show()

print(f'Area solicitada = {c:.4f}')

print(f'El valor critico solicitado es k ={d}')
print(f'P(X<={d})={e:.4f}')
print(f'P({d}=<X<={a})={b-e:.4f}')
print(f'P(X<={a})={b:.4f}')

```
