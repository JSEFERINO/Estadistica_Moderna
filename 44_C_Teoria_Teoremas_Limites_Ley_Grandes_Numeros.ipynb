# **196. Teoremas Límites - Convergencia**
# **a. Definición - Tipos de Convergencia**
+ Sea $X_1,X_2,⋯,X_n,⋯$, una sucesión de variables aleatorias.
+ **Definición** Si para cada $ω∈Ω$ la sucesión numérica $X_1(ω),X_2(ω),⋯,X_n(ω),⋯$ es convergente, entonces
 - (a) Se dice que la sucesión $X_1,X_2,⋯,X_n,⋯$ converge **puntualmente**
  - (b) $X(ω)=\lim_{n→∞}X_n(ω)$ es una variable aleatoria

# **b. Como medir la cercanía de $X_n$ con $X$**
+ (1) **Convergencia casi segura**
$$P(X_n→X)=1$$

+ (2) **Convergencia en Media**
$$E|X_n-X|→0$$

+ (3) **Convergencia en Probabilidad**
$$P(|X_n-X|>\in)→0$$

+ (4) **Convergencia en Distribución**
$$F_{X_n}(x)→F_X(x)$$

# **c. ¿A dónde converge $\bar X=\frac{1}{n}\sum_{1}^{n}X_i$?**
+ (1) **Ley de los Grandes Números**
$$\frac{1}{n}\sum_{1}^{n}X_i→μ$$


+ (2) **Teorema Central del Límite - TCL**
$$\frac{\frac{1}{n}\sum_{1}^{n}X_i-μ}{\sqrt{\frac{σ^2}{n}}}→Z∼\mathbb{N}(0,1)$$


  
# **197. Ley de los grandes números**
# **a. Generalidades**

- En la teoría de la probabilidad, bajo el término genérico de ley de los grandes números se engloban varios teoremas que describen el comportamiento del promedio de una sucesión de variables aleatorias conforme aumenta su número de ensayos.

- Estos teoremas prescriben condiciones suficientes para garantizar que dicho promedio converja (en los sentidos explicados abajo) al promedio de las esperanzas de las variables aleatorias involucradas.

- Las distintas formulaciones de la ley de los grandes números (y sus condiciones asociadas) especifican la convergencia de formas distintas.

- Las leyes de los grandes números explican por qué el promedio de una muestra al azar de una población de gran tamaño tenderá a estar cerca de la media de la población completa.

- Cuando las variables aleatorias tienen una varianza finita, el teorema central del límite extiende nuestro entendimiento de la convergencia de su promedio describiendo la distribución de diferencias estandarizadas entre la suma de variables aleatorias y el valor esperado de esta suma: sin importar la distribución subyacente de las variables aleatorias, esta diferencia estandarizada converge a una variable aleatoria normal estándar.

# **b. Ley débil**
- La ley débil de los grandes números establece que si ${\displaystyle X_{1},X_{2},X_{3},\dots }$ es una sucesión infinita de variables aleatorias independientes que tienen el mismo valor esperado ${\displaystyle \mu }$ y varianza ${\displaystyle \sigma ^{2}}$ entonces el promedio

$${\displaystyle {\begin{aligned}{\overline {X}}_{n}&={\frac {X_{1}+\cdots +X_{n}}{n}}\\&={\frac {1}{n}}\sum _{i=1}^{n}X_{i}\end{aligned}}}$$
converge en probabilidad a $\mu$, en otras palabras, para cualquier número positivo $\varepsilon$  se tiene

$${\displaystyle \lim _{n\rightarrow \infty }\operatorname {P} \left(\left|{\overline {X}}_{n}-\mu \right|<\varepsilon \right)=1.}$$


# **c. Ley fuerte**
- La ley fuerte de los grandes números establece que si ${\displaystyle X_{1},X_{2},X_{3},\dots }$ es una sucesión infinita de variables aleatorias independientes e idénticamente distribuidas que cumplen ${\displaystyle \operatorname {E} [X_{i}]<\infty }$ y tienen el valor esperado ${\displaystyle \operatorname {E} [X_{i}]=\mu }$ entonces

$${\displaystyle \operatorname {P} \left(\lim _{n\rightarrow \infty }{\overline {X}}_{n}=\mu \right)=1,}$$
 - es decir, el promedio de las variables aleatorias converge a ${\displaystyle \mu }$ casi seguramente (en un conjunto de probabilidad 1).

+ Esta ley justifica la interpretación intuitiva del valor esperado de una variable aleatoria como el "promedio a largo plazo al hacer un muestreo repetitivo".



# **198. Ejemplos de la Ley de los grandes Números**
# **a. Ejemplo 1. Lanzamiento de una Moneda - La ley de los grandes números**
+ Sea $X_1,X_2,⋯,X_n,⋯$, una sucesión de variables aleatorias de Bernoulli independientes.
+ $P\{X_i=1\}=P\{Cara\}=p$, $P\{X_i=0\}=P\{Sello\}=1-p$
+ Es claro que la Variable aleatoria $$Y=\sum_{1}^{n}X_i∼\mathbb{B}(n,p)$$
+ Por la ley de los grandes números
$$\bar X=\frac{1}{n}\sum_{1}^{n}X_i →p$$ cuando $n→∞$

+ **En otras palabras La ley de los grandes números dice que la probabilidad de que un evento ocurra es igual a su frecuencia relativa**
+ El Evento Sale cara $A=\{X_i=1\}$ la probabilidad que ocurra en $n$ lanzamientos es $$P(A)=\lim_{n→∞}\frac{\sum_{1}^{n}X_i}{n}=\frac{Casos_{favorables}}{Total_{casos}}=p=0.5$$

```python
# **a. Ejemplo 1. Lanzamiento de una moneda - La ley de grandes números**
# moneda p=P{X=1}=P{'Sale Cara'} = 1/2,  p=P{X=0}=P{'Sale Sello'} = 1/2

import random # Libreria a usar

promedios = [] # Almacenaaremos los promedios de los lanzamientos

for lanzamientos in range(1,10000): # Vamos hacer 10000 lanzamientos
    lanzamientos = np.random.choice([0,1], lanzamientos,  p=[1/2,1/2]) # En los lanzamientos solo vamos a obtener {0, 1}
    exitos = lanzamientos.mean() # Vamos a hallar el promedio de caras o 1
    promedios.append(exitos) # almacenaremos este resultado arriba

# graficamente
df = pd.DataFrame({ 'promedios' : promedios}) # Construimos el Data Frame

df.plot(title='Ley de grandes números',color='b',figsize=(8, 6)) # Nos hara el gráfico de los resultados
plt.axhline(0.5) # esta es la linea correspondiente a p=P{X=1}=P{'Sale Cara'} = 1/2
plt.xlabel("Número de lanzamientos") #Nombre del eje X
plt.ylabel("Probabilidad de salir exitos") # Nombre del eje Y
plt.show()

print('Un resumen de la data obtenida',df.describe())
df.value_counts()
```

# **b. Ejemplo 2. Lanzamiento de un Dado -  La ley de los grandes números**
+ Sea $X_1,X_2,⋯,X_n,⋯$, una sucesión de variables aleatorias de Bernoulli independientes
+ $\{X_i=1\}=\{éxito\}$, $\{X_i=0\}=\{fracaso\}$
+ Dado que  $$p=P({X_i=j)}=1/6, j=1,2,3,4,5,6$$

+ $\{X_i=1\}=\{Cara\}$, $\{X_i=0\}=\{Sello\}$
+ Es claro que la Variable aleatoria $$Y=\sum_{1}^{n}X_i∼\mathbb{B}(n,p)$$
+ Por la ley de los grandes números
$$\bar X=\frac{1}{n}\sum_{1}^{n}X_i →p$$ cuando $n→∞$


```python
# **b. Ejemplo 2. Lanzamiento de un dado - La ley de grandes números**
# Dado p=P{X=1}=P{X=2}=P{X=3}=P{X=4}=P{X=5}=P{X=6}= 1/6
# Dado p=P{exito}= 1/6

import random # Libreria a usar

promedios = [] # Almacenaaremos los promedios de los lanzamientos

for lanzamientos in range(1,10000): # Vamos hacer 10000 lanzamientos
    lanzamientos = np.random.choice([0,1], lanzamientos, p=[5/6,1/6]) # En los lanzamientos solo vamos a obtener {0, 1}
    exitos = lanzamientos.mean() # Vamos a hallar el promedio de caras o 1
    promedios.append(exitos) # almacenaremos este resultado arriba

# graficamente
df = pd.DataFrame({ 'promedios' : promedios}) # Construimos el Data Frame

df.plot(title='Ley de grandes números',color='b',figsize=(8, 6)) # Nos hara el gráfico de los resultados
plt.axhline(1/6) # esta es la linea correspondiente a p=P{X=1}=P{'Sale Cara'} = 1/2
plt.xlabel("Número de lanzamientos") #Nombre del eje X
plt.ylabel("Probabilidad de salir exito") # Nombre del eje Y
plt.show()

print('Un resumen de la data obtenida',df.describe())
df.value_counts()
```

# **c. Ejemplo 3. Lanzamiento de un Dado -  La ley de los grandes números - Otra interpretación**
+ Supongamos que lanzamos un dado $n$ veces, La Ley de los grandes números Afirma que $$P\{X_i=j\}=1/6$$ en la medida que  $n→∞$
+ **Veamos en la siguiente Simulación usando python**


```python
# **c. Ejemplo: Simular el lanzamiento de un dado varias veces - Usando Python**
import random
import seaborn.objects as so

a = 1 # Valor minimo en un dado
b = 6 # Valor máximo en un dado
x=1 # nos va indicar el primer lanzamiento
dado = [] # Queremos colocar el resultado obtenido en los lanzamientos


while x<=1000000: # Esto es para decir que queremos lanzar el dado 10 veces
    valor=random.randint(a,b) # Cada vez que se henere un lanzamiento, se genera un numero aleatorio entre 1 y 6
    dado.append(valor) # Me dice el resultado del dato


    x=x+1 # el ciclo se cierra y comienza

dados = pd.DataFrame({'dado':dado}) # Construimos el DataFrame de los resultados

dados.value_counts(normalize = True)
print('Tablas de frecuencias relativas')
print(dados.value_counts(normalize = True))
print('Histograma de frecuencias relativas')
so.Plot(dados, x="dado", color="dado").add(so.Bar(), so.Hist(stat="proportion"))
```
