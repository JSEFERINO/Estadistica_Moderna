# **174. Modelos Probabilísticos Continuos - La distribución Log_Normal**

# **a. Generalidades de la Distribución Log_Normal**
- Los siguientes fenómenos se pueden modelar con una distribución lognormal:
+ $X=:$ Producción de leche por vacas.
+ $X=:$ Vidas de unidades industriales con modos de falla que se caracterizan por fatiga-esfuerzo.
+ $X=:$ Cantidades de lluvia.
+ $X=:$ Distribuciones de tamaño de las gotas de lluvia.
+ $X=:$ El volumen de gas en una reserva de petróleo.
+ $X=:$ Duración de los períodos latentes de enfermedades infecciosas
+ $X=:$ Abundancia de especies
+ En la **hidrología**, se utiliza la distribución log-normal para analizar variables aleatorias como valores máximos de la precipitación y la descarga de ríos,​ y además para describir épocas de sequía.

+ **Otros nombres para el lognormal.**Históricamente, la distribución lognormal ha recibido muchos nombres, entre ellos:

 - La distribución de Galton o Galton (después del estadístico victoriano Francis Galton).
 - Distribución de Galton-McAlister (McAlister fue otro estadístico victoriano que publicó una descripción de la distribución con Galton).
 - Distribución de Gibrat , en honor al economista francés del siglo XX que demostró que los logaritmos de ciertas variables económicas (como la distribución de fábricas por número de trabajadores) seguían una distribución normal.
 - Distribución Cobb-Douglas (después de los economistas del siglo XX Charles Cobb y Paul Douglas). *Este término se usa exclusivamente en economía, donde se aplica a los datos de producción.*



# **b. Distribuciones continuas - La Distribución Log - Normal**

+ Una variable aleatoria continua $X$ sigue una distribución Log - Normal de parámetros $μ$ y $σ$, lo cual denotamos por $$X∼LN(μ , σ)$$ si tiene funcion FDP dada por la formula:
$$ P(X=x)= f(x,μ , σ) = \frac{1}{x\sqrt{2\piσ^2}}exp{\{\frac{-1}{2σ^2}\}(logx-μ)^2}, \ x>0$$
+ Tambien $$E(X)=\mu=e^{μ+σ^2/2},  \ \ \ \ V(X)=e^{2μ+σ^2}(e^{σ^2}-1)$$

+ En dónde $ \{X\leq x\}$ representa al evento de todos resultados menores que $x$.
+ $\mu$ representa al número esperado.
+ $V(X)$ representa a la varianza.
+ Se puede mostrar


# **c. Valor Esperado de una Variable aleatoria normal**

+  $$E(X)=\mu$$
+ $$E(X) = ∫_{-∞}^{∞}xf(x)dx= ∫_{-∞}^{∞}\frac{x}{x\sqrt{2\piσ^2}}exp{[\frac{-1}{2σ^2}](lnx-μ)^2}dx$$

+ El cambio de variable $ω = \frac{lnx-μ}{σ}$
+ donde $x=e^{σω+μ}$ y $dx=σe^{σω+μ}dω$


+ $$E(X) = ∫_{-∞}^{∞}\frac{1}{\sqrt{2\piσ^2}}exp{[\frac{-ω^2}{2}]}σe^{σω+μ}dω$$

+ $$E(X) = e^{μ}∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-ω^2}{2}]}e^{σω}dω$$

+ $$E(X) = e^{μ}∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-ω^2+2σω}{2}]}dω$$
+ $$E(X) = e^{μ}∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-ω^2+2σω-σ^2+σ^2}{2}]}dω$$


+ **Separando términos en las integrales**
+ $$E(X) = e^{μ+σ^2/2}∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-ω^2+2σω-σ^2}{2}]}dω$$

+ $$E(X) = e^{μ+σ^2/2}∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-(ω-σ)^2}{2}]}dω$$
+ $$E(X) = e^{μ+σ^2/2}∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-z^2}{2}]}dω=e^{μ+σ^2/2}$$


# **d. Varianza de una Variable aleatoria normal**

+  $$V(X)=σ^2$$
+ $$E(X^2) = ∫_{-∞}^{∞}x^2f(x)dx= ∫_{-∞}^{∞}\frac{x^2}{x\sqrt{2\piσ^2}}exp{[\frac{-1}{2σ^2}](lnx-μ)^2}dx$$

+ El cambio de variable $ω = \frac{lnx-μ}{σ}$
+ donde $x=e^{σω+μ}$ y $dx=σe^{σω+μ}dω$

+ $$E(X^2)  = ∫_{-∞}^{∞}\frac{e^{σω+μ}}{\sqrt{2\pi}}exp{[\frac{-ω^2}{2}]}e^{σω+μ}dω$$

+ **Reorganizando la integral**

+ $$E(X^2)  = e^{2μ} ∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-ω^2+4σω}{2}]}dω$$

+ $$E(X^2)  = e^{2μ+2σ^2} ∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-ω^2+4σω-4σ^2}{2}]}dω$$

+ $$E(X^2)  = e^{2μ+2σ^2} ∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-(ω-2σ)^2}{2}]}dω = e^{2μ+2σ^2} $$


+ **Se tiene**


+ $$V(X)= E(X^2)-(E(X))^2 =e^{2μ+2σ^2} -(e^{μ+σ^2/2})^2 $$
+ $$V(X)= E(X^2)-(E(X))^2 =e^{2μ+2σ^2} -(e^{2μ+σ^2})=e^{2μ}{e^{2σ^2}} -e^{2μ}{e^{σ^2}}=e^{2μ}({e^{2σ^2}} -{e^{σ^2}})=e^{2μ+σ^2}({e^{σ^2}} -1) $$

#**a. Construya el gráfico de la Distribución dada si los parámetros son $μ=5,σ=1$**

```python
# **a. Grafico manual de la función
from numpy.ma.core import log
# nuestra libreria
import numpy as np
import matplotlib.pyplot as plt
# Definimos una funcion
media = 5 # parametro
sigma = 1 # parametro
def lognormal1(x):
  return np.exp(-((log(x)-media)**2)/(2*sigma**2))/(x*np.sqrt(2*3.1416*sigma**2))
# los valores que va a tomar x
x = np.linspace(0,2000,100) # espacio donde vamos a graficar la funcion
fig, ax = plt.subplots() #lugar donde vamos a estar graficando, ax objeto y subplot porque vamos hacer varios graficos
ax.plot(x,lognormal1(x)) # variable independiente y la funcion
plt.title('Distribución LogNormal')
ax.grid() # le da mas elegancia al grafico
```

```python
#**a. Construya el gráfico de la Distribución dada si los parámetros son μ=5,σ=1**
plt.figure(figsize=(8, 8))
lognorm = stats.lognorm(s=1, scale=np.exp(5))
x = np.linspace(lognorm.ppf(0.01),
                lognorm.ppf(0.99), 100)
fp = lognorm.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución LogNormal')
plt.ylabel('Probabilidad')
plt.xlabel('Ingresos')
plt.show()
```

#**b. Construya el histograma para una muestra aleatoria tomada de una población lognormal  si los parámetros son $\sigma=1$ y $\mu=5$**

```python
#**b. Construya el histograma para una muestra aleatoria tomada de una población lognormal  si los parámetros son $\sigma=1$ y $\mu=5$**
aleatorios = stats.lognorm(s=1, scale=np.exp(5)).rvs(10000) # genera aleatorios
cuenta, cajas, ignorar = plt.hist(aleatorios, 100)

plt.ylabel('frequencia')
plt.xlabel('Ingresos')
plt.title('Histograma Distribución LogNormal con $μ=5,σ=1$')
plt.show()
```

# **i. ¿Qué porcentaje de los ingresos de los habitantes se encuentra entre 150 y 600**

```python
# **i. ¿Qué porcentaje de los ingresos de los habitantes se encuentra entre 150 y 600**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(0, 1500, 401)

y = stats.lognorm(s=1, scale=np.exp(5)).pdf(x)
a = 150 # Valor de la variable a calcular a la izquierda
b = 600 # Valor de la variable a calcular a la izquierda
c = stats.lognorm(s=1, scale=np.exp(5)).cdf(b)-stats.lognorm(s=1, scale=np.exp(5)).cdf(a) # Probabilidad pedida
x1 = np.linspace(a, b, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.lognorm(s=1, scale=np.exp(5)).pdf(x1) , color="r")
#ax.fill_between(x[x<87], y[x<87], color="b")

ax.text(350, 0.001, f'{c:.4f}', horizontalalignment='center')
#ax.text(95, 0.05, f'{1 - stats.norm(mu, sd).cdf(a):.4f}', horizontalalignment='center')

ax.set_xlabel('Ingresos')
ax.set_ylabel('Densidad')
plt.title('Distribución Normal con $μ=5,σ=1$')
ax.text(a, 0.00001, a)
ax.text(b, 0.00001, b)
ax.text(450, 0.001, f'Area solicitada = {c:.4f}')
print('la probabilidad de que una batería dure entre 150 y 600 es')
print('P{150=<X<=600} =',f'Area solicitada = {c:.4f}')

plt.show()
```

# **m. Dada una distribución lognormal  con $μ=5,σ=1$, calcule el valor de k tal que P (k < X < 550) = 0.42**

```python
# **m. Dada una distribución lognormal  con $μ=5,σ=1$, calcule el valor de k tal que P (k < X < 550) = 0.42**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(0, 1500, 401)

y = stats.lognorm(s=1, scale=np.exp(5)).pdf(x)
a = 550 # Valor de la variable a calcular a la izquierda
b = stats.lognorm(s=1, scale=np.exp(5)).cdf(a)
c = stats.lognorm(s=1, scale=np.exp(5)).cdf(a) - 0.42
d = round(stats.lognorm(s=1, scale=np.exp(5)).ppf(c),1) # valor critico pedido
e = stats.lognorm(s=1, scale=np.exp(5)).cdf(d) # Probabilidad acumulada

x1 = np.linspace(0, d, 401)
x2 = np.linspace(d, a, 401)
x3 = np.linspace(a, 1500, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.lognorm(s=1, scale=np.exp(5)).pdf(x1) , color="r", alpha = 0.3)
ax.fill_between(x2, 0,stats.lognorm(s=1, scale=np.exp(5)).pdf(x2) , color="b", alpha =0.3)
ax.fill_between(x3, 0,stats.lognorm(s=1, scale=np.exp(5)).pdf(x3) , color="g", alpha =0.3)

ax.text(d, 0.00001, d)
ax.text(a, 0.00001, a)
ax.text(d-50, 0.001, f'Area solicitada = {e:.4f}')
ax.text(d+100, 0.0005, f'Area dada = {0.42:.4f}')

ax.set_xlabel('Valores de x')
ax.set_ylabel('Densidad')
plt.title('Distribución Normal con $μ=5,σ=1$')
print('El área a la izquierda de 550 es P{X<=550} = ',b)
print('El área a la izquierda de k es es P{X<=k}=' ,c)
print('El valor critico solicitado es k =',d)
print(f'Confirmamos que k es el valor buscado con probabilidad P(X<={d})',e)

plt.show()
```
