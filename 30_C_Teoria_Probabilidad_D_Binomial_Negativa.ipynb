# **164. La Distribución Binomial Negativa**
## **a. La Defincición**
+ Una variable aleatoria discreta $X$ sigue una distribución Binomial Negativa de parámetros $r$ y $p$, lo cual denotamos por $$X∼B_{neg}(r,p)$$ si tiene funcion FMP dada por la formula:
$$ P(X=x)= f(x,r,p) = \begin{pmatrix} x+r-1 \\ r-1 \end{pmatrix}p^r(1-p)^{x},  x=0, 1, 2, ..., $$


+ Donde $$\begin{pmatrix} x+r-1 \\ r-1 \end{pmatrix}=\begin{pmatrix} x+r-1 \\ x \end{pmatrix}=\frac{(x+r-1)!}{(r-1)!x!}$$

+ Tambien $$E(X)=\mu=\frac{r(1-p)}{p},  \ \ \ \  V(X)=σ^2=\frac{r(1-p)}{p^2}$$

+ En dónde $ X=x$ representa el número de fracasos que se producen antes de obtener el $r-éxito$.
+ $\mu$ representa al número esperado de fracasos en todo el experimento.
+ $V(X)=σ^2$ representa a la variabilidad presente durante el experimento aleatorio.


## **b. Valor Esperado y Varianza de una variable aleatoria discreta de Binomial negativa**

+ **Recordemos que $P(X=x)$ esta bien definida y esto nos serivra para concluir mas abajo**
+ $$ \sum_{x=o}^{∞}P(X=x) =\sum_{x=o}^{∞}\begin{pmatrix} x+r-1 \\ x \end{pmatrix}p^x(1-p)^x=\sum_{x=o}^{∞}\frac{(x+r-1)!}{(r-1)!x!}p^r(1-p)^x=1$$

* ****
+ **Valor Esperado** $$E(X)=\mu=\sum_{x=o}^{∞}x\begin{pmatrix} r+x-1 \\ x \end{pmatrix}p^r(1-p)^x$$

* ****
+ **Se tiene que**
+ $$E(X)=\sum_{x=0}^{∞} x{\frac{(x+r-1)!}{x!(r-1)!}p^r(1-p)^x}$$

+ $$E(X)=\sum_{x=0}^{∞} x{\frac{(x+r-1)!}{x(x-1)!(r-1)!}p^r(1-p)^x}$$


+ $$E(X)=\sum_{x=1}^{∞} {\frac{(x+r-1)!}{(x-1)!(r-1)!}p^r(1-p)^x}$$

* ****

+ **Reorganizando**

+ $$E(X)=\sum_{x=1}^{∞} {\frac{((x-1)+(r+1)-1)!}{(x-1)!(r-1)!}p^r(1-p)^x}$$
* ****
+ **Multiplicando y dividindo por $r$**
+ $$E(X)=r\sum_{x=1}^{∞} {\frac{((x-1)+(r+1)-1)!}{(x-1)!r(r-1)!}p^r(1-p)^x}$$

+ ****
+ $$E(X)=r\sum_{x=1}^{∞} {\frac{((x-1)+(r+1)-1)!}{(x-1)!r!}p^r(1-p)^x}$$
* ****
+ **Usando la definición del coeficiente binomial**
+ $$E(X)=r\sum_{x=1}^{∞} {\begin{pmatrix} (x-1)+(r+1)-1 \\ x-1 \end{pmatrix}p^r(1-p)^x}$$

* ****
+ **Reacomodando las potencias de $p$ y $1-p$**

+ $$E(X)=r\sum_{x=1}^{∞} {\begin{pmatrix} (x-1)+(r+1)-1 \\ x-1 \end{pmatrix}p^{(r+1)}p^{-1}(1-p)^{x-1}(1-p)}$$

* ****

+ **Reacomodando todo**

+ $$E(X)=\frac{r(1-p)}{p}\sum_{x=1}^{∞} {\begin{pmatrix} (x-1)+(r+1)-1 \\ x-1 \end{pmatrix}p^{(r+1)}(1-p)^{x-1}}$$

* ****

+ **Reacomodando todo con $y=x-1$ y $R=r+1$**

+ $$E(X)=\frac{r(1-p)}{p}\sum_{y=0}^{∞} {\begin{pmatrix} y+R-1 \\ y \end{pmatrix}p^R(1-p)^y}$$

* ****

+ **Aplicamos lo visto en la parte arriba**

+ $$E(X)=\frac{r(1-p)}{p}[1]=\frac{r(1-p)}{p}$$

* ****


## **c. Varianza**
+ $$V(X)=E(X-μ)^2=σ^2=E(X^2)- {\mu}^2$$

+ **Recordemos que $P(X=x)$ esta bien definida y esto nos servirá para concluir mas abajo**
+ $$ \sum_{x=o}^{∞}P(X=x) =\sum_{x=o}^{∞}\begin{pmatrix} x+r-1 \\ x \end{pmatrix}p^x(1-p)^x=\sum_{x=o}^{∞}\frac{(x+r-1)!}{(r-1)!x!}p^r(1-p)^x=1$$

* ****

+ **Calculamos la $E(X^2)$**
$$E(X^2)=E(X(X-1)+X)=E(X(X-1))+E(X)$$ \\

$$E(X(X-1))=\sum_{x=o}^{∞}x(x-1)\begin{pmatrix} r+x-1 \\ x \end{pmatrix}p^r(1-p)^x  $$



+ **Dado que**


$$E(X(X-1))=\sum_{x=o}^{∞}x(x-1)\frac{(x+r-1)!}{(r-1)!x!}p^r(1-p)^x$$


$$E(X(X-1))=\sum_{x=o}^{∞}x(x-1)\frac{(x+r-1)!}{(r-1)!x(x-1)(x-2)!}p^r(1-p)^x$$


$$E(X(X-1))=\sum_{x=2}^{∞}\frac{(x+r-1)!}{(r-1)!(x-2)!}p^r(1-p)^x$$

+ **Reacomodando**


$$E(X(X-1))=\sum_{x=2}^{∞}\frac{((x-2)+(r+2)-1)!}{(r-1)!(x-2)!}p^r(1-p)^x$$

+ ****
+ $$E(X(X-1))=r(r+1)\sum_{x=2}^{∞}\frac{((x-2)+(r+2)-1)!}{(r+1)!(x-2)!}p^r(1-p)^x$$


* ****
+ **Reacomodando las potencias de $p$ y $1-p$**

+ $$E(X(X-1))=r(r+1)\sum_{x=2}^{∞}\frac{((x-2)+(r+2)-1)!}{(r+1)!(x-2)!}p^{r+2}p^{-2}(1-p)^{x-2}(1-p)^2$$

* ****

+ **Reacomodando todo**

+ $$E(X(X-1))=\frac{r(r+1)(1-p)^2}{p^2}\sum_{x=2}^{∞}\frac{((x-2)+(r+2)-1)!}{(r+1)!(x-2)!}p^{r+2}(1-p)^{x-2}$$


* ****

+ **Reacomodando todo con $y=x-2$ y $R=r+2$**

+ $$E(X(X-1))=\frac{r(r+1)(1-p)^2}{p^2}\sum_{y=0}^{∞}\frac{(y+R-1)!}{(R-1)!y!}p^R(1-p)^y$$


* ****

+ **Aplicamos lo visto en la parte arriba**

+ $$E(X(X-1))=\frac{r(r+1)(1-p)^2}{p^2}[1]=\frac{r(r+1)(1-p)^2}{p^2}$$

* ****
+ **Aplicamos lo visto en la parte arriba para $E(X^2)$**

$$E(X^2)=E(X(X-1))+E(X)$$

$$E(X^2)=\frac{r(r+1)(1-p)^2}{p^2}+\frac{r(1-p)}{p}$$

+ **Varianza**
+ $$V(X)=E(X-μ)^2=σ^2=E(X^2)- {\mu}^2$$

+ $$V(X)=\frac{r(r+1)(1-p)^2}{p^2}+\frac{r(1-p)}{p}-[\frac{r(1-p)}{p}]^2$$

+ **Factorizando**

+ $$V(X)=[\frac{r(1-p)}{p^2}][(r+1)(1-p)+p-r(1-p)]=\frac{r(1-p)}{p^2}$$

# **a. Construya el gráfico de la Distribución dada si los parámetros son $r=5,p=0.8$**

```python
# **a. Graficando la Distribución Binomial negativa**
plt.figure(figsize=(8, 6))
r, p = 5, 0.8 # parametros de forma
nbinom = stats.nbinom(r, p) # Distribución
x = np.arange(nbinom.ppf(0.01),
              nbinom.ppf(0.99))
fmp = nbinom.pmf(x) # Función de Masa de Probabilidad
plt.plot(x, fmp, 'b')
plt.vlines(x, 0, fmp, colors='r', lw=5, alpha=0.5)
plt.title('Distribución Binomial Negativa (r = 5, p =0.8)')
plt.ylabel('probabilidad')
plt.xlabel('X = El número de reacciones negativas antes de que apareciera la 5ta reacción positiva')
plt.show()
```

# **b. PMF y FDA la Distribución Binomial Negativa**

```python
# **b. PMF y FDA la Distribución Binomial Negativa**
x=list(range(0,11)) #El número de reacciones negativas antes de que apareciera la 5ta reacción positiva
rv=stats.nbinom(5, 0.8)
df=pd.DataFrame({'i':x,'Probabilidad P{X=i}':rv.pmf(x),'Distribución F(i)=P{X<=i}':rv.cdf(x),})
df
round(df,3)
```


# **d. Suponga que se tienen 3 tipos de exámenes para detectar una cierta enfermedad con probabilidades respectivas de generar una reacción positiva de 0.25, 0.35 y 0.40, Construya en un solo cuadro las distribuciones que representan cada caso**

```python
# **d. Comparaciones Distribución Binomial Negativa**
x = np.linspace(0, 50, 51)
distributions = pd.DataFrame({
    'x': x,
    'dist_p=0.25': stats.nbinom(5, 0.25).pmf(x),
    'dist_p=0.35': stats.nbinom(5, 0.35).pmf(x),
    'dist_p=0.40': stats.nbinom(5, 0.4).pmf(x),
})
ax = distributions.plot(x='x', y='dist_p=0.25', legend=True, color='r')
distributions.plot(x='x', y='dist_p=0.35', legend=True, ax=ax, color='b', linestyle='--')
distributions.plot(x='x', y='dist_p=0.40', legend=True, ax=ax, color='g', linestyle='-.')
ax.set_ylim(0, 0.14)

plt.title('Función de Masa de Probabilidad Binomial Negativa con r = 5')
plt.ylabel('probabilidad')
plt.xlabel('X = El número de reacciones negativas antes de que apareciera la 5ta reacción positiva')
ax.get_legend()
plt.show()
```

# **k. Grafique la Función Generadora de Momentos y la  Función Generadora de Probabilidad**
+ **Función Generadora de Momentos** $$M_X(t)=E(e^{tX})=(\frac{p}{(1-(1-p)e^t)}{)^r}$$

+ **Función Generadora de Probabilidad** $$G_X(t)=E(t^X)=(\frac{p}{(1-(1-p)t)}{)^r}$$

```python
# **k. Grafique la Función Generadora de Momentos y la  Función Generadora de Probabilidad**
plt.figure(figsize=(25, 10))
p = 0.8
r = 5
x=np.linspace(0,1,1000, dtype = "float")
generadoras=pd.DataFrame({'x':x,'G_Momentos':(p/(1-(1-p)*2.71**x))**r,'G_Probabilidad':(p/(1-(1-p)*x))**r})
ax=generadoras.plot(x='x',y='G_Momentos',color='r')
generadoras.plot(x='x',y='G_Probabilidad',color='b',ls='--',ax=ax)

ax.set_ylim(0, 10)
ax.text(0.3,6,'G_Momentos')
ax.text(0.5,0.5,'G_Probabilidad')

ax.get_legend()
plt.ylabel('observados en y')
plt.xlabel('valores de x')
plt.title('G_Momentos vs G_Probabilidad ')
plt.show()
```
