## **8.2 Definición clásica (Laplace)**.
- La probabilidad de un suceso $A$ es el cociente entre el número de casos favorables al suceso y el número de casos posibles.

$$ P(A)=\frac{\sharp (A)}{\sharp (\Omega)} $$
- Los inconvenientes de definir la probabilidad de esta forma son:*No es válida cuando los sucesos elementales no son equiprobables*, *A veces no es posible contar*.

## **8.3 Definición frecuentista (Bernouilli)**
- La probabilidad de un suceso es el valorlímite de su frecuencia relativa al repetir indefinidamente la experimentación.

- Sea $\Omega$ un espacio muestral y $\mathcal{F}$ una $\sigma$-álgebra de subconjuntos de $\Omega$.
\item La probabilidad de un evento $A \in \mathcal{F}$, está dada por el límite

$$
\mathbb{P}(A) = \lim_{n \rightarrow \infty} \dfrac{n_A}{n}
$$
- en donde $n$ es el número total de ensayos y $n_A$ es el número de veces que ocurre el evento $A$.
- Los inconvenientes de definir así la probabilidad son los siguientes:
- Desde el punto de vista del análisis no puede interpretarse el límite anterior por la imposibilidad de fijar el número de repeticiones.
- En algunas ocasiones no es posible realizar una experimentación indefinida.
- Las condiciones bajo las cuales se realiza la experimentación pueden variar a lo largo del tiempo y, con ellas, las frecuencias relativas.

- En la práctica, la definición de probabilidad frecuentista se encuentra limitada debido al hecho de que en la realidad $n$ y $n_A$ son números finitos y por lo tanto, lo que realmente se tiene es la aproximación

$$
\mathbb{P}(A) \approx \lim_{n \rightarrow \infty} \dfrac{n_A}{n}
$$

# **Ejemplo: Simulación de lanzamiento de moneda $n$ veces**
## **$\bullet$** **Este experimento simula el lanzamiento de una moneda el numero de veces que queramos**

```python
import random

def lanzamiento_moneda():
    resultado = random.choice(['cara', 'cruz'])
    return resultado

# Simulación de lanzar la moneda 5 veces
for i in range(10):
    print("Lanzamiento", i+1, ":", lanzamiento_moneda())
```

```python
import matplotlib.pyplot as plt

# Función para simular lanzamientos de moneda y obtener resultados
def lanzamiento_moneda(n):
    resultados = [random.choice(['cara', 'cruz']) for _ in range(n)]
    return resultados

# Simulación de lanzar la moneda 100 veces
resultados = lanzamiento_moneda(100)

# Contar la frecuencia de cada resultado
cara_count = resultados.count('cara')
cruz_count = resultados.count('cruz')

!pip install texttable &>/dev/null # La usaremos para crea tablas
from texttable import Texttable # La usaremos para crea tablas

# Tabla Resumen
t1 = Texttable()
t1.add_rows([['Resultado','frecuencia', 'Probabilidad' ],
           ['cara',	cara_count, cara_count/(cara_count+cruz_count)],
           ['cruz',	cruz_count, cruz_count/(cara_count+cruz_count)]])
print(t1.draw())

# Generar histograma
plt.bar(['Cara', 'Cruz'], [cara_count, cruz_count], color = ['blue', 'salmon'])
plt.title('Histograma de Lanzamiento de Moneda')
plt.xlabel('Resultado')
plt.ylabel('Frecuencia')
plt.show()
```

# **Ejemplo: Simulación de lanzamiento  $n$ veces de un dado**
## **$\bullet$** **Probabilidad Limite**

```python
import matplotlib.pyplot as plt

# Function for simulating dice rolls and obtaining results
def lanzamiento_dado(n):
    resultados = [random.choice(['uno', 'dos','tres', 'cuatro','cinco', 'seis',]) for _ in range(n)]
    return resultados

# Simulation of rolling the dice 100 times
resultados = lanzamiento_dado(100000)

# Count the frequency of each outcome
uno_count = resultados.count('uno')
dos_count = resultados.count('dos')
tres_count = resultados.count('tres')
cuatro_count = resultados.count('cuatro')
cinco_count = resultados.count('cinco')
seis_count = resultados.count('seis')

# Calculate the total number of outcomes
Total = uno_count + dos_count + tres_count + cuatro_count + cinco_count + seis_count

# Create a table to summarize the results
t1 = Texttable()
t1.add_rows([['Resultado','frecuencia', 'Probabilidad' ],
           ['uno',	uno_count, uno_count/Total],
           ['dos',	dos_count, dos_count/Total],
           ['tres',	tres_count, tres_count/Total],
           ['cuatro',	cuatro_count, cuatro_count/Total],
           ['cinco',	cinco_count, cinco_count/Total],
           ['seis',	seis_count, seis_count/Total]])
print(t1.draw())

# Generar histograma
plt.bar(['uno', 'dos','tres', 'cuatro','cinco', 'seis'], [uno_count/Total, dos_count/Total,tres_count/Total,cuatro_count/Total,cinco_count/Total,seis_count/Total ])
plt.title('Histograma de Lanzamiento de Dado')
plt.xlabel('Resultado')
plt.ylabel('Probabilidad')
plt.show()
```
