# **b. Recordemos cuales son nuestras columnas**

# **c. Tablas Pivote - Datos Numericos**
+ Una variable cualitativa con dos variables cuantitativas

```python
# **c. Tablas Pivote - Datos Numericos**
# Para lashortalizas se realizan las ganancias y los costos
datos.pivot_table(index = 'Sistema', values=['Costo', 'Ganancia'], aggfunc=np.sum)
```

# **b. Datos marginales -Para las hortalizas se realizan las ganancias y los costos**

```python
# **b. Datos marginales -Para las hortalizas se realizan las ganancias y los costos**
datos.pivot_table(index = 'Sistema', values=['Costo', 'Ganancia'], margins=True, aggfunc=np.sum)
```

# **c. Promedios Datos marginales -Para las hortalizas se realizan las ganancias y los costos**

```python
# **c. Promedios Datos marginales -Para las hortalizas se realizan las ganancias y los costos**
datos.pivot_table(index = 'Sistema', values=['Costo', 'Ganancia'], margins=True, aggfunc=np.mean)
```

# **d. Redondeo de Promedios Datos marginales -Para las hortalizas se realizan las ganancias y los costos**

```python
# **d. Redondeo de Promedios Datos marginales -Para las hortalizas se realizan las ganancias y los costos**
round(datos.pivot_table(index = 'Sistema', values=['Costo', 'Ganancia'], margins=True, aggfunc=np.mean),2)
```

# **e. Dos variables cualitativas y una cuantitativa**

```python
# **e. Dos variables cualitativas y una cuantitativa**
datos.pivot_table(index = ['Sistema','Hortaliza'], values=['Costo'], margins=True, aggfunc=np.sum)
```

# **f. Tres variables cualitativas y una cuantitativa**

```python
# **f. Tres variables cualitativas y una cuantitativa**
datos.pivot_table(index = ['Sistema','Hortaliza'], values=['Costo'], columns =['pH'], margins=True, aggfunc=np.sum)
```

# **g. Tres variables cualitativas y una cuantitativa - Elimanos la posición sin información**

```python
# **g. Tres variables cualitativas y una cuantitativa - Elimanos la posición sin información**
datos.pivot_table(index = ['Sistema','Hortaliza'], values=['Costo'], columns =['pH'], margins=True, aggfunc=np.sum,
                  fill_value=0)
```

# **h. Multiples variables cualitativas y una cuantitativa**

```python
# **h. Multiples variables cualitativas y una cuantitativa**
datos.pivot_table(index = 'Hortaliza', values=['Costo'], columns =['Sistema'],
                  aggfunc={'Costo':[np.sum, np.mean],'Sistema': len},
                  fill_value=0)
```

# **i. Multiples variables cualitativas y una cuantitativa - Redondeo**

```python
# **i. Multiples variables cualitativas y una cuantitativa - Redondeo**
round(datos.pivot_table(index = 'Hortaliza', values=['Costo'], columns =['Sistema'],
                  aggfunc={'Costo':[np.sum, np.mean],'Sistema': len},
                  fill_value=0),2)
```
