#  **6.2. Propiedades de los Eventos**.

$${A \cup A = A}$$ \\
$${A \cup \varnothing= A}$$ \\
  $${A \cup \Omega= \Omega}$$

##***Intersección***
$${A \cup \overline{A}= \Omega}$$
$${A \cap B= B \cap A}$$

$${A \cap \varnothing= \varnothing}$$

$${A \cap \overline{A}= \varnothing}$$

$${A - B= A - (A \cap B)}$$

#  **6.4. Conjunto potencia**.

- Sea $\Omega \not =\varnothing$ un conjunto.
- El conjunto que contiene a todos los subconjunto $A$ de $\Omega$ se llama conjunto potencia y se denota $$\wp(\Omega)=2^\Omega =\{A \mid A\subseteq \Omega\},$$

- Si $\Omega<\infty$, se cumple que $$\sharp(2^\Omega)=2^{\sharp(\Omega)}$$


###  **Ejemplo**:
- Si $$\Omega  =\{a,b,c \}, \\ $$ $$2^\Omega =\{\varnothing, \{a\}, \{b\}, \{c\}, \{a,b\},\{a,c\},\{b,c\}, \Omega\}$$ \\
- $$\sharp(2^\Omega)= 2^3 =8$$



# **6.5 Producto cartesiano entre dos conjuntos**.

- El producto cartesiano entre dos conjuntos $A$ y $B$ viene dado por  $$A \times B =\{(x,y) \mid x \in A, y \in B \}$$

###  **a. Ejemplo** Sean $A =\{0,1\}$ y $B =\{x,y\}$. Entonces,
+ $$A \times B =\{(0,x),(0,y),(1,x),(1,y) \}$$
- Análogamente se define $${A_1\times A_2\times\cdots\times A_n}$$

###  **b. Propiedades del producto cartesiano**
- $$A \times B \not =B \times A$$ \\
- Si $\sharp(A) =m$ y $\sharp(B) =n$, entonces,  $$\sharp (A \times B ) =mn$$ \\
- $$\sharp(A \times B) =\sharp(B \times A)$$ \\
- $${\sharp({A_1\times A_2\times\cdots\times A_k)=n_1 n_2\cdots n_k}}$$



###  **c. Ejemplo**

- $${\mathbb{R}^{n} =\mathbb{R}\times\cdots\times \mathbb{R}=\{(x_1,\cdots,x_n) \mid x_1 \in \mathbb{R},\dots, x_n \in \mathbb{R} \}}$$

- $A_1,A_2,\dots,A_n$ son mutuamente excluyentes dos a dos si ${A_i\cap A_j=\varnothing}$, para ${i\not=j;i,j=1,\dots,n}$
- Una colección infinita de eventos $A_1,A_2,\dots,A_n,\dots,$ son *mutuamente excluyentes dos a dos* si cualquier subconjunto finito de ella tambien lo es.

#  **7.2 Sigma álgebra**

#####  **Definción**
- Una colección, $\mathcal{F}$, no vacía de subconjuntos de un conjunto $\Omega$, se llama una *sigma álgebra* ($\sigma$-álgebra) si cumple las siguientes propiedades:

- $\Omega \in \mathcal{F}$.
- Si $A \in \mathcal{F}$, entonces $A^{c} \in \mathcal{F}$.
- Si $A_n \in \mathcal{F}$, $n = 1,2, \ldots$, entonces $$\displaystyle\bigcup_{n=1}^{\infty}A_n \in \mathcal{F}$$ y $$\displaystyle \bigcap_{n=1}^{\infty}A_n \in \mathcal{F}$$.
- De acuerdo a la definición, podemos restringir nuestra atención únicamente a los eventos que pertenecen a una sigma álgebra.



###  **Ejemplo de sigma- álgebra: σ-álgebra generada por una colección de conjuntos**: _Dada una colección arbitraria de conjuntos, puedes construir una σ-álgebra generada por esos conjuntos. Esta σ-álgebra contendrá todos los subconjuntos que se pueden formar a partir de la colección original y será la más pequeña que los contenga._
- Si $$\Omega  =\{a,b,c \}, \\ $$ $$2^\Omega =\{\varnothing, \{a\}, \{b\}, \{c\}, \{a,b\},\{a,c\},\{b,c\}, \Omega\}$$ \\
- $$\sharp(2^\Omega)= 2^3 =8$$


```python
class SigmaAlgebra:
    def __init__(self, sample_space):
        self.sample_space = list(sample_space)  # Convert the set to a list
        self.events = [set()]

    def add_event(self, event):
        if event.issubset(self.sample_space):
            self.events.append(event)
        else:
            print("El evento no es un subconjunto del espacio muestral.")

    def generate_power_set(self):
        self.events = [set()]
        for i in range(len(self.sample_space)):
            new_events = [event.union({self.sample_space[i]}) for event in self.events]
            self.events.extend(new_events)

    def print_sigma_algebra(self):
        for event in self.events:
            print(event)

    def count_elements(self):
        return len(self.events)


# Ejemplo de uso
sample_space = {'a', 'b', 'c'}  # Espacio muestral
sigma = SigmaAlgebra(sample_space)
sigma.generate_power_set()
sigma.print_sigma_algebra()

# Contar el número de elementos en la sigma-álgebra
num_elements = sigma.count_elements()
print("Número de elementos en la sigma-álgebra:", num_elements)
```

##  **7.3 Otros Ejemplos de σ-álgebras**.

###  **a. Ejemplo. σ-álgebra de Borel en la recta real**: En $\mathbb{R}$ (la recta real), la ***σ-álgebra de Borel*** es una σ-álgebra que contiene todos los conjuntos abiertos, cerrados y semiabiertos. También contiene todos los intervalos abiertos, semicerrados y cerrados. Es una σ-álgebra fundamental en análisis real y teoría de la medida.

+ Una sigma-álgebra de Borel es una sigma-álgebra particularmente importante en la teoría de la medida y la teoría de la probabilidad. Se define sobre un espacio topológico y contiene todos los conjuntos que se pueden formar a partir de los conjuntos abiertos (y sus complementos) mediante operaciones como uniones, intersecciones y complementos.

+ Aquí tienes un ejemplo de cómo construir una sigma-álgebra de Borel sobre la recta real utilizando Python:


```python
class BorelSigmaAlgebra:
    def __init__(self, real_line):
        self.real_line = real_line

    def generate_borel_sigma_algebra(self):
        open_sets = [(a, b) for a in self.real_line for b in self.real_line if a < b]  # Conjuntos abiertos
        closed_sets = [(-float('inf'), a) for a in self.real_line] + [(a, float('inf')) for a in self.real_line]  # Conjuntos cerrados
        compact_sets = [(a, b) for a in self.real_line for b in self.real_line if a < b and a != -float('inf') and b != float('inf')]  # Conjuntos compactos

        borel_sets = set(open_sets + closed_sets + compact_sets)

        self.borel_sigma_algebra = set()

        for borel_set in borel_sets:
            complements = [(-float('inf'), borel_set[0]), (borel_set[1], float('inf'))]
            for complement in complements:
                self.borel_sigma_algebra.add(borel_set)
                self.borel_sigma_algebra.add(complement)
                intersection = (max(borel_set[0], complement[0]), min(borel_set[1], complement[1]))
                if intersection[0] < intersection[1]:
                    self.borel_sigma_algebra.add(intersection)

    def print_borel_sigma_algebra(self):
        for borel_set in self.borel_sigma_algebra:
            print(borel_set)


# Ejemplo de uso
real_line = range(-5, 6)  # Recta real
borel_sigma = BorelSigmaAlgebra(real_line)
borel_sigma.generate_borel_sigma_algebra()
borel_sigma.print_borel_sigma_algebra()
```

+ Este ejemplo construye una sigma-álgebra de Borel sobre la recta real con los conjuntos abiertos, cerrados y compactos como base. Luego, utiliza las operaciones de unión, intersección y complemento para generar todos los conjuntos en la sigma-álgebra de Borel.

+ Ten en cuenta que este es solo un ejemplo simplificado y no incluye todos los conjuntos de la sigma-álgebra de Borel. La recta real se ha representado con un conjunto discreto de números para facilitar la visualización. En una implementación real, podrías trabajar con números reales y utilizar métodos más sofisticados para generar la sigma-álgebra de Borel.

###  **b. Ejemplo. σ-álgebra discreta**: En un conjunto finito o infinito numerable, la σ-álgebra discreta contiene todos los subconjuntos posibles. Si tienes un conjunto $X = \{x_1,x_2,...,x_n\}$, la σ-álgebra discreta consiste en todos los subconjuntos posibles de $X$, incluyendo el conjunto vacío y $X$ mismo.


```python
class DiscreteSigmaAlgebra:
    def __init__(self, sample_space):
        self.sample_space = sample_space
        self.sigma_algebra = self.generate_discrete_sigma_algebra()

    def generate_discrete_sigma_algebra(self):
        power_set = []
        for i in range(len(self.sample_space) + 1):
            power_set.extend(combinations(self.sample_space, i))
        return power_set

    def print_sigma_algebra(self):
        for event in self.sigma_algebra:
            print(event)


# Ejemplo de uso
from itertools import combinations

sample_space = {'A', 'B', 'C'}  # Espacio muestral discreto
discrete_sigma = DiscreteSigmaAlgebra(sample_space)
discrete_sigma.print_sigma_algebra()
```
