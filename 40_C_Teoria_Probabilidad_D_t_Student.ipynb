# **187. Distribuciones continuas - La Distribución *t* de Student**
# **a. Generalidades de la Distribución *t* de Student**
+ **Wikipedia**. En probabilidad y estadística, la distribución $t$ de Student es una distribución de probabilidad que surge del problema de estimar la media de una población normalmente distribuida cuando el tamaño de la muestra es pequeño y la desviación estándar poblacional es desconocida.

- Fue desarrollada por William Sealy Gosset bajo el pseudónimo “Student”.
- Aparece de manera natural al realizar la prueba t de Student para la determinación de las diferencias entre dos varianzas muestrales y para la construcción del intervalo de confianza para la diferencia entre las partes de dos poblaciones cuando se desconoce la desviación típica de una población y esta debe ser estimada a partir de los datos de una muestra.


- *Historia*. La distribución de Student fue descrita en el año 1908 por *William Sealy Gosset*. Gosset trabajaba en una fábrica de cerveza, *Guinness*, que prohibía a sus empleados la publicación de artículos científicos debido a una difusión previa de secretos industriales. De ahí que Gosset publicase sus resultados bajo el pseudónimo de *“Student”*




#  **b. Distribución t de Student a partir de una muestra aleatoria**
- Sea ${\displaystyle X_{1},\cdots ,X_{n}}$ variables aleatorias independientes distribuidas ${\displaystyle N(\mu ,\sigma ^{2})}$, esto es, ${\displaystyle X_{1},\dots ,X_{n}}$ es una muestra aleatoria de tamaño ${\displaystyle n}$ proveniente de una población con distribución normal con media $\mu$ y varianza $\sigma ^{2}$.

- **Consideremos los Estadísticos: la media muestral y la varianza muestral**
$$
{\displaystyle {\overline {X}}={\frac {1}{n}}\sum _{i=1}^{n}X_{i}}$$


$${\displaystyle S^{2}={\frac {1}{n-1}}\sum _{i=1}^{n}\left(X_{i}-{\overline {X}}\right)^{2}}$$

- **Entonces, la variable aleatoria**

$${\displaystyle {\frac {{\overline {X}}-\mu }{\sigma /{\sqrt {n}}}}}$$
- sigue una distribución normal estándar (es decir, una distribución normal con media 0 y varianza 1)
- y la variable aleatoria

$${\displaystyle {\frac {{\overline {X}}-\mu }{S/{\sqrt {n}}}}}$$
- donde ${\displaystyle S}$ ha sido sustituido por ${\displaystyle \sigma }$, tiene una distribución $t$ de student con $n-1$ grados de libertad.




# **c. Función de densidad  t de Student**
- Sean $X$ una variable aleatoria continua y ${\displaystyle n>0}$, si ${\displaystyle X}$ tiene una distribución $t$ con $v$ grados de libertad entonces escribiremos ${\displaystyle X\sim t_{n}}$ o ${\displaystyle X\sim t(n)}$ si tiene  **función de densidad de probabilidad**
$${\displaystyle f_{X}(x)={\frac {\Gamma \left({\frac {n+1}{2}}\right)}{{\sqrt {v\pi }}\;\Gamma \left({\frac {n}{2}}\right)}}\left(1+{\frac {x^{2}}{n}}\right)^{-{\frac {n+1}{2}}}}$$
para ${\displaystyle x\in \mathbb {R} }$, donde ${\displaystyle v}$ denota los grados de libertad y ${\displaystyle \Gamma }$ es la función gamma.

- **La expresión anterior también suele escribirse como:**
$$f_{X}(x)={\frac {1}{{\sqrt {n}}\;\operatorname {B} \left({\frac {1}{2}},{\frac {n}{2}}\right)}}\left(1+{\frac {x^{2}}{n}}\right)^{-{\frac {n+1}{2}}}$$
donde $B$ es la función beta.


+ **d. Propiedades**

- Si ${\displaystyle X}$ es una variable aleatoria tal que ${\displaystyle X\sim t_{n}}$ entonces $X$ satisface:

- **Media**. La media de $X$ para valores ${\displaystyle n>1}$ es
$${\displaystyle \operatorname {E} [X]=0}$$
- **Varianza**. La varianza de $X$ para valores ${\displaystyle n>2}$ es

$${\displaystyle \operatorname {Var} (X)={\frac {n}{n-2}}}$$



# **d. Caracterización**
+ La distribución $t$ de Student con $v$ grados de libertad puede definirse como la distribución de la variable aleatoria $T$ definida por:

$${\displaystyle T:={\frac {Z}{\sqrt {\frac {X}{v}}}}\sim t_{v}}$$
- donde ${\displaystyle Z\sim N(0,1)}$, es decir, ${\displaystyle Z}$ es una variable aleatoria con distribución normal estándar (distribución normal con media 0 y varianza 1).

- ${\displaystyle X\sim \chi _{v}^{2}}$, es decir $X$ es una variable aleatoria que sigue una distribución chi-cuadrada con $v$ grados de libertad.
- $Z$ y $X$  son variables aleatorias independientes.
- Para una constante {\displaystyle \mu }\mu no nula, el cociente

- ${\displaystyle (Z+\mu ){\sqrt {\frac {v}{X}}}}$ es una variable aleatoria que sigue la distribución no central $t$  de Student con parámetro de no-centralidad $\mu$.

# **e. Intervalo para la media**
- $\sigma^2$ es desconocida

- Sean ${\displaystyle X_{1},\dots ,X_{n}}$ una muestra aleatoria proveniente de una población con distribución $N(\mu ,\sigma ^{2})$ donde $\mu$ y $\sigma$ son desconocidos.

- Se tiene que
$${\displaystyle {\frac {{\overline {X}}-\mu }{\sigma /{\sqrt {n}}}}\sim N(0,1)}$$
- y
$${\displaystyle {\frac {(n-1)S^{2}}{\sigma ^{2}}}\sim \chi _{n-1}^{2}}$$
- son independientes entonces el cociente
$${\displaystyle {\frac {\frac {{\overline {X}}-\mu }{\sigma /{\sqrt {n}}}}{\sqrt {\frac {\frac {(n-1)S^{2}}{\sigma ^{2}}}{n-1}}}}\sim t_{n-1}}$$
- esto es
$${\displaystyle {\frac {{\overline {X}}-\mu }{S/{\sqrt {n}}}}\sim t_{n-1}}$$
- Sea ${\displaystyle t_{n-1,1-\alpha /2}\in \mathbb {R} }$ tal que

- ${\displaystyle \operatorname {P} [Y\leq t_{n-1,1-\alpha /2}]=1-{\frac {\alpha }{2}}}$ siendo ${\displaystyle Y\sim t_{n-1}}$ entonces

 $${\displaystyle {\begin{aligned}&\operatorname {P} \left[-t_{n-1,1-\alpha /2}\leq {\frac {{\overline {X}}-\mu }{S/{\sqrt {n}}}}\leq t_{n-1,1-\alpha /2}\right]=1-\alpha \\&\operatorname {P} \left[-t_{n-1,1-\alpha /2}\;{\frac {S}{\sqrt {n}}}\leq {\overline {X}}-\mu \leq t_{n-1,1-\alpha /2}\;{\frac {S}{\sqrt {n}}}\right]=1-\alpha \\&\operatorname {P} \left[-{\overline {X}}-t_{n-1,1-\alpha /2}\;{\frac {S}{\sqrt {n}}}\leq -\mu \leq -{\overline {X}}+t_{n-1,1-\alpha /2}\;{\frac {S}{\sqrt {n}}}\right]=1-\alpha \\&\operatorname {P} \left[{\overline {X}}-t_{n-1,1-\alpha /2}\;{\frac {S}{\sqrt {n}}}\leq \mu \leq {\overline {X}}+t_{n-1,1-\alpha /2}\;{\frac {S}{\sqrt {n}}}\right]=1-\alpha \\\end{aligned}}}$$
- por lo tanto un intervalo de ${\displaystyle (1-\alpha )100\%}$ de confianza para $\mu$ cuando $\sigma ^{2}$  es desconocida es

$${\displaystyle \left({\overline {X}}-t_{n-1,1-\alpha /2}\;{\frac {S}{\sqrt {n}}},{\overline {X}}+t_{n-1,1-\alpha /2}\;{\frac {S}{\sqrt {n}}}\right)}$$

# **d. Veamos que es una función de densidad**

+ **Densidad Distribución t de Student**
 $$f_X(x)=\frac{Γ(\frac{n+1}{2})}{\sqrt {n\pi }Γ(\frac{n}{2})}(1+\frac{x^2}{n})^{-\frac{n+1}{2}}$$

+ **Integrando**.
$$ ∫_{-∞}^{∞}f(x)dx= ∫_{-∞}^{∞}\frac{Γ(\frac{n+1}{2})}{\sqrt {n\pi }Γ(\frac{n}{2})}(1+\frac{x^2}{n})^{-\frac{n+1}{2}}dx$$
$$ ∫_{-∞}^{∞}f(x)dx= 2\frac{Γ(\frac{n+1}{2})}{\sqrt {n\pi }Γ(\frac{n}{2})}∫_{0}^{∞}(1+\frac{x^2}{n})^{-\frac{n+1}{2}}dx$$


+ **hacemos** $y = (1+\frac{x^2}{n})^{-1}$ esto es $x = \sqrt {n}(y^{-1}-1)^{1/2} ⇒dx=-\frac{1}{2} \sqrt {n}(1-y)^{-1/2}y^{-3/2}dy$

+ **Sustituyendo en la Integral**.
$$ ∫_{-∞}^{∞}f(x)dx= -2\frac{Γ(\frac{n+1}{2})}{\sqrt {n\pi }Γ(\frac{n}{2})}∫_{1}^{0}(y)^{\frac{n+1}{2}}\frac{1}{2} \sqrt {n}(1-y)^{-1/2}y^{-3/2}dy$$

+ **Reorganizando la Integral**.
$$ ∫_{-∞}^{∞}f(x)dx= \frac{Γ(\frac{n+1}{2})}{\sqrt {\pi }Γ(\frac{n}{2})}∫_{0}^{1}y^{\frac{n}{2}-1} (1-y)^{\frac{1}{2}-1}dy$$

+ **Recordando la función Beta**

 $$
{B(α,β)} = \int_0^{1} x^{α-1}(1-x)^{β-1}dx=\frac{\Gamma(α)\Gamma(β)}{\Gamma(α+β)}
$$

+ **Reemplazando en la  Integral**.
$$ ∫_{-∞}^{∞}f(x)dx= \frac{Γ(\frac{n+1}{2})}{\sqrt {\pi }Γ(\frac{n}{2})}∫_{0}^{1}y^{\frac{n}{2}-1} (1-y)^{\frac{1}{2}-1}dy=\frac{Γ(\frac{n+1}{2})}{\sqrt {\pi }Γ(\frac{n}{2})}B(\frac{n}{2},\frac{1}{2})$$

+ **Usando la propiedad de la funcion Beta arriba**.
$$ ∫_{-∞}^{∞}f(x)dx= \frac{Γ(\frac{n+1}{2})}{\sqrt {\pi }Γ(\frac{n}{2})}\frac{\Gamma(\frac{n}{2})\Gamma(\frac{1}{2})}{\Gamma(\frac{n+1}{2})}=1$$


# **e. Valor Esperado y varianza**

+ **Valor Esperado**.
$$ E(X) = ∫_{-∞}^{∞}xf(x)dx= ∫_{-∞}^{∞}\frac{Γ(\frac{n+1}{2})}{\sqrt {n\pi }Γ(\frac{n}{2})}x(1+\frac{x^2}{n})^{-\frac{n+1}{2}}dx = 0$$

+ **Valor Esperado de $X^2$**.
$$ E(X^2) = ∫_{-∞}^{∞}x^2f(x)dx= ∫_{-∞}^{∞}\frac{Γ(\frac{n+1}{2})}{\sqrt {n\pi }Γ(\frac{n}{2})}x^2(1+\frac{x^2}{n})^{-\frac{n+1}{2}}dx = 0$$
$$ E(X^2) =∫_{-∞}^{∞}x^2f(x)dx= 2\frac{Γ(\frac{n+1}{2})}{\sqrt {n\pi }Γ(\frac{n}{2})}∫_{0}^{∞}x^2(1+\frac{x^2}{n})^{-\frac{n+1}{2}}dx$$

+ **hacemos** $y = (1+\frac{x^2}{n})^{-1}$ esto es $x = \sqrt {n}(y^{-1}-1)^{1/2}  ⇒dx=-\frac{1}{2} \sqrt {n}(1-y)^{-1/2}y^{-3/2}dy$

 - **Donde** $$x^2 = n(1-y)y^{-1}$$

+ **Sustituyendo en la Integral**.
$$ E(X^2) == -2\frac{Γ(\frac{n+1}{2})}{\sqrt {n\pi }Γ(\frac{n}{2})}∫_{1}^{0} n(1-y)y^{-1}(y)^{\frac{n+1}{2}}\frac{1}{2} \sqrt {n}(1-y)^{-1/2}y^{-3/2}dy$$


+ **Reorganizando en la Integral**.
$$ E(X^2) = n\frac{Γ(\frac{n+1}{2})}{\sqrt {\pi }Γ(\frac{n}{2})}∫_{0}^{1} (1-y)y^{-1}(y)^{\frac{n+1}{2}} (1-y)^{-1/2}y^{-3/2}dy$$

$$E(X^2) = n\frac{Γ(\frac{n+1}{2})}{\sqrt {\pi }Γ(\frac{n}{2})}∫_{0}^{1} y^{\frac{n-2}{2}-1}(1-y)^{3/2-1}dy$$

+ **Recordando la función Beta**

 $$
{B(α,β)} = \int_0^{1} x^{α-1}(1-x)^{β-1}dx=\frac{\Gamma(α)\Gamma(β)}{\Gamma(α+β)}
$$

+ **Usando la función Beta**

$$ E(X^2) =n\frac{Γ(\frac{n+1}{2})}{\sqrt {\pi }Γ(\frac{n}{2})}B(\frac{n-2}{2},\frac{3}{2})$$

+ **Usando la función Beta**
$$ E(X^2) = nB(\frac{n}{2},\frac{1}{2})B(\frac{n-2}{2},\frac{3}{2})$$



+ **c. Propiedades de la función Beta y gamma**
+ $Γ(α+1)=αΓ(α)$
+ $B(α,β)=B(β,α)$
+ $B(α,1)=\frac{\Gamma(α)\Gamma(1)}{\Gamma(α+1)}=\frac{\Gamma(α)(1)}{α\Gamma(α)}=\frac{1}{α}$

+ $B(α+1,β)=\frac{\Gamma(α+1)\Gamma(β)}{\Gamma(α+β+1)}=\frac{α\Gamma(α)\Gamma(β)}{(α+β)\Gamma(α+β)}=\frac{α}{(α+β)}B(α,β)$


+ **Usando las propiedades de la función Beta**

$$ E(X^2) =n\frac{Γ(\frac{n+1}{2})}{\sqrt {\pi }Γ(\frac{n}{2})}\frac{Γ(\frac{n-2}{2})Γ(\frac{3}{2})}{Γ(\frac{n+1}{2})}$$

 -  $Γ(\frac{3}{2})=Γ(\frac{1}{2}+1)=\frac{1}{2}Γ(\frac{1}{2})=\frac{1}{2}\sqrt {\pi }$

 + **Usando las propiedades de la función Gamma y simplificando**

$$ E(X^2) =\frac{n}{2}\frac{1}{Γ(\frac{n}{2})}\frac{Γ(\frac{n-2}{2})}{1}$$
-  $Γ(\frac{n}{2})=Γ(\frac{n-2}{2}+1)=\frac{n-2}{2}Γ(\frac{n-2}{2})$

+ **Usando las propiedades de la función Gamma y simplificando**

$$ E(X^2) = \frac{n}{2}\frac{1}{\frac{n-2}{2}Γ(\frac{n-2}{2})}\frac{Γ(\frac{n-2}{2})}{1}=\frac{n}{n-2}$$

+ **Por tanto**
$$Var(X)=E(X^2)- E(X)^2 = =\frac{n}{n-2}$$

#**a. Construya el gráfico de la Distribución dada**
+ **Densidad Distribución t - Student**

$$f_X(x)=\frac{Γ(\frac{n+1}{2})}{\sqrt {n\pi }Γ(\frac{n}{2})}(1+\frac{x^2}{n})^{-\frac{n+1}{2}}$$


$$f_X(x)=q_0(1+\frac{x^2}{n})^{-\frac{n+1}{2}}$$

```python

#**a. Construya el gráfico de la Distribución dada si  $α=50$ y $β=49$**

# Definimos una funcion
n = 25 # parametro alpha
a = n/2
b = (n+1)/2

gamma_a = math.gamma(a)
gamma_b = math.gamma(b)
c = sqrt(n*np.pi)
q_0 = gamma_b/(c*gamma_a)

def dis_t(x):
  return q_0*(1+x**2/n)**(-b)
# los valores que va a tomar x
x = np.linspace(-5,5,1000)
fig, ax = plt.subplots() #lugar donde vamos a estar graficando, ax objeto y subplot porque vamos hacer varios graficos
ax.plot(x,dis_t(x)) # variable independiente y la funcion

plt.title('Fabrica A - Distribución t Stdent')
plt.ylabel('Probabilidad')
plt.xlabel('X :contenido faltante de las bolsas')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

```python
#**a. Construya el gráfico de la Distribución dada si  α=11.7 y β=7.24**
plt.figure(figsize=(8, 8))
df = 25 # parametro de forma.
t = stats.t(df)

x = np.linspace(t.ppf(0.0001),
                t.ppf(0.99999), 10000)
fp = t.pdf(x) # Función de Probabilidad
plt.plot(x, fp)

plt.title('Fabrica A - Distribución t Stdent')
plt.ylabel('Probabilidad')
plt.xlabel('X :contenido faltante de las bolsas')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

#  **k. Dada la Fábrica A, calcule el valor de $k$  tal que $P (k < X < 1.5) = 0.42$**

```python
# **k**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(-5, 5, 401)

y = stats.t(25).pdf(x)
a = 1.0 # Valor de la variable a calcular a la izquierda
b = stats.t(25).cdf(a)
c = stats.t(25).cdf(a) - 0.42
d = round(stats.t(25).ppf(c),2) # valor critico pedido
e = stats.t(25).cdf(d) # Probabilidad acumulada
x1 = np.linspace(-5, d, 401)
x2 = np.linspace(d, a, 401)
x3 = np.linspace(a, 5, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.t(25).pdf(x1) , color="r", alpha = 0.3)
ax.fill_between(x2, 0,stats.t(25).pdf(x2) , color="b", alpha =0.3)
ax.fill_between(x3, 0,stats.t(25).pdf(x3) , color="y", alpha =0.3)

ax.text(d, 0.00001, d)
ax.text(a, 0.00001, a)

ax.text(a, 0.1, f'Area1=P(X>{a}) = {1-b:.4f}')
ax.text((a+d)/2, 0.05, f'Area2=P({d}=<X<={a})= {c:.4f}')
ax.text(d-4, 0.15, f'Area3= P(X<={d}) = {e:.4f}')


print(f'Area solicitada = {c:.4f}')

print(f'El valor critico solicitado es k ={d}')
print(f'P(X<={d})={e:.4f}')
print(f'P({d}=<X<={a})={b-e:.4f}')
print(f'P(X<={a})={b:.4f}')


plt.title('Fabrica A - Distribución t Stdent')
plt.ylabel('Probabilidad')
plt.xlabel('X :contenido faltante de las bolsas')
ax.grid() # le da mas elegancia al grafico
plt.show()
```
