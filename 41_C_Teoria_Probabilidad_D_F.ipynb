# **189. Distribuciones continuas - La Distribución $F$**
# **a. Generalidades de la Distribución $F$**
+ **Wikipedia**. En teoría de probabilidad y estadística, la distribución F, también conocida como distribución de Fisher-Snedecor (nombrada por Ronald Fisher y George Snedecor), es una distribución de probabilidad continua, aparece frecuentemente como la distribución nula de una prueba estadística, especialmente en el análisis de varianza.
+ La distribución F es una distribución continua de muestreo de la relación de dos variables aleatorias independientes con distribuciones de chi-cuadrada, cada una dividida entre sus grados de libertad. La distribución F es asimétrica hacia la derecha y es descrita por los grados de libertad de su numerador (ν1) y denominador (ν2). Las siguientes gráficas muestran el efecto de los diferentes valores de grados de libertad en la forma de la distribución



# **b. La Distribución F - Definición**

+ Una variable aleatoria continua $X$ sigue una distribución *F* de parámetros $m$ y $n$, lo cual denotamos por $$X∼F_{(m,n)}$$ si tiene funcion FDP dada por la formula:
$$ f(x,m,n) = \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}(\frac{m}{n})^{\frac{m}{2}}\frac{x^{\frac{m-2}{2}}}{(1+\frac{mx}{n})^{\frac{m+n}{2}}}, \ x>0$$


+ Tambien $$E(X)=\mu=\frac{n}{n-2}  \ \ \ \ V(X)=σ^2=\frac{2n^2(m+n-2)}{m(n-2)^2(n-4)}$$



# **c. Veamos que es una función de densidad**

+ **Densidad Distribución F**
$$ f(x,m,n) = \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}(\frac{m}{n})^{\frac{m}{2}}\frac{x^{\frac{m-2}{2}}}{(1+\frac{mx}{n})^{\frac{m+n}{2}}}, \ x>0$$

+ **Integrando**.
$$ ∫_{-∞}^{∞}f(x)dx= ∫_{0}^{∞}\frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}(\frac{m}{n})^{\frac{m}{2}}\frac{x^{\frac{m-2}{2}}}{(1+\frac{mx}{n})^{\frac{m+n}{2}}}dx$$


$$ ∫_{-∞}^{∞}f(x)dx= \ \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}(\frac{m}{n})^{\frac{m}{2}} ∫_{0}^{∞}\frac{x^{\frac{m-2}{2}}}{(1+\frac{mx}{n})^{\frac{m+n}{2}}}dx$$

***
  
+ **hacemos** $$y = (1+\frac{mx}{n})^{-1}$$ esto es $$x =\frac{n}{m}y^{-1}(1-y) ⇒dx=-\frac{n}{m}y^{-2}dy$$
+
$$x^{\frac{m-2}{2}} =(\frac{n}{m}y^{-1}(1-y))^{\frac{m-2}{2}}=(\frac{n}{m})^{\frac{m-2}{2}}(y^{-{\frac{m-2}{2}}})(1-y)^{\frac{m-2}{2}}$$
***
+ **Sustituyendo en la Integral**.

$$ ∫_{-∞}^{∞}f(x)dx= - \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}(\frac{m}{n})^{\frac{m}{2}} ∫_{1}^{0}\frac{(\frac{n}{m})^{\frac{m-2}{2}}(y^{-{\frac{m-2}{2}}})(1-y)^{\frac{m-2}{2}}}{y^{-\frac{m+n}{2}}}\frac{n}{m}y^{-2}dy$$
***
+ **Reorganizando la Integral**.

$$ ∫_{-∞}^{∞}f(x)dx=  \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}∫_{0}^{1}\frac{(y^{-{\frac{m-2}{2}}})(1-y)^{\frac{m-2}{2}}}{y^{-\frac{m+n}{2}}}y^{-2}dy$$

$$ ∫_{-∞}^{∞}f(x)dx=  \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}∫_{0}^{1}y^{{\frac{n}{2}}-1}(1-y)^{{\frac{m}{2}}-1}dy$$


+ **Recordando la función Beta**

 $$
{B(α,β)} = \int_0^{1} x^{α-1}(1-x)^{β-1}dx=\frac{\Gamma(α)\Gamma(β)}{\Gamma(α+β)}
$$

+ **Reemplazando en la  Integral**.
$$ ∫_{-∞}^{∞}f(x)dx=  \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}∫_{0}^{1}y^{{\frac{n}{2}}-1}(1-y)^{{\frac{m}{2}}-1}dy=\frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}\frac{Γ(\frac{m}{2})Γ(\frac{n}{2})}{Γ(\frac{m+n}{2})}=1$$

# **c. Varianza**

+ **Densidad Distribución F**
$$ f(x,m,n) = \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}(\frac{m}{n})^{\frac{m}{2}}\frac{x^{\frac{m-2}{2}}}{(1+\frac{mx}{n})^{\frac{m+n}{2}}}, \ x>0$$

+ **Integrando**.
$$ E(X^2)=∫_{-∞}^{∞}x^2f(x)dx= ∫_{0}^{∞}\frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}(\frac{m}{n})^{\frac{m}{2}}\frac{x^{\frac{m-2}{2}+2}}{(1+\frac{mx}{n})^{\frac{m+n}{2}}}dx$$


$$ E(X^2)=∫_{-∞}^{∞}xf(x)dx= \ \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}(\frac{m}{n})^{\frac{m}{2}} ∫_{0}^{∞}\frac{x^{\frac{m}{2}+1}}{(1+\frac{mx}{n})^{\frac{m+n}{2}}}dx$$

***
  
+ **hacemos** $$y = (1+\frac{mx}{n})^{-1}$$ esto es $$x =\frac{n}{m}y^{-1}(1-y) ⇒dx=-\frac{n}{m}y^{-2}dy$$
+
$$x^{\frac{m}{2}+1} =(\frac{n}{m}y^{-1}(1-y))^{\frac{m}{2}+1}=(\frac{n}{m})^{\frac{m}{2}+1}(y^{-{\frac{m}{2}}-1})(1-y)^{\frac{m}{2}+1}$$
***
+ **Sustituyendo en la Integral**.

$$ ∫_{-∞}^{∞}x^2f(x)dx= - \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}(\frac{m}{n})^{\frac{m}{2}} ∫_{1}^{0}\frac{(\frac{n}{m})^{\frac{m}{2}+1}(y^{-{\frac{m}{2}}-1})(1-y)^{\frac{m}{2}+1}}{y^{-\frac{m+n}{2}}}\frac{n}{m}y^{-2}dy$$
***
+ **Reorganizando la Integral**.

$$ ∫_{-∞}^{∞}x^2f(x)dx=  \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}(\frac{n}{m})^2∫_{0}^{1}y^{({\frac{n}{2}}-2)-1}(1-y)^{({\frac{m}{2}}+2)-1}$$


+ **Recordando la función Beta**

 $$
{B(α,β)} = \int_0^{1} x^{α-1}(1-x)^{β-1}dx=\frac{\Gamma(α)\Gamma(β)}{\Gamma(α+β)}
$$

+ **Reemplazando en la  Integral**.
$$ ∫_{-∞}^{∞}x^2f(x)dx=  \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}(\frac{n}{m})^2\frac{Γ(\frac{n}{2}-2)Γ(\frac{m}{2}+2)}{Γ(\frac{m+n}{2})}$$

+ **Sabemos que**
$$Γ(α)=(α-1)Γ(α-1)=(α-1)(α-2)Γ(α-2)$$
+ **Cambiando $α=\frac{m}{2}+2$**
$$Γ(\frac{m}{2}+2)=(\frac{m}{2}+1)\frac{m}{2}Γ(\frac{m}{2})$$
+ **Cambiando $α=\frac{2}{2}$**
$$Γ(\frac{n}{2})=(\frac{n}{2}-1)(\frac{n}{2}-2)Γ(\frac{n}{2}-2)$$

+ **Reemplazando y simplificando la integral arriba**

$$ E(X^2)=∫_{-∞}^{∞}x^2f(x)dx= (\frac{n}{m})^2 \frac{(m+2)m}{(n-2)(n-4)}$$
+ **Por lo tanto**
$$Var(X)=σ^2=\frac{n^2}{m} \frac{2(m+n-2)}{(n-2)^2(n-4)}$$

#**a. Construya el gráfico de la Distribución dada**
+ **Densidad Distribución t - Student**


+ **Densidad Distribución F**
$$ f(x,m,n) = \frac{Γ(\frac{m+n}{2})}{Γ(\frac{m}{2})Γ(\frac{n}{2})}(\frac{m}{n})^{\frac{m}{2}}\frac{x^{\frac{m-2}{2}}}{(1+\frac{mx}{n})^{\frac{m+n}{2}}}, \ x>0$$


$$f_X(x)=q_0\frac{x^{\frac{m-2}{2}}}{(1+\frac{mx}{n})^{\frac{m+n}{2}}}$$

```python

#**a.**

# Definimos una funcion
m = 20 # parametro alpha
n = 25

a = ((m/n)**(m/2))*math.gamma((m+n)/2)
b = math.gamma(n/2)*math.gamma(m/2)
c = (m+n)/2
d = (m-2)/2

q_0 = a/b

def dis_F(x):
  return q_0*(x**d)/(((1+m*x/n)**c))
# los valores que va a tomar x
x = np.linspace(0,8,1000)
fig, ax = plt.subplots() #lugar donde vamos a estar graficando, ax objeto y subplot porque vamos hacer varios graficos
ax.plot(x,dis_F(x)) # variable independiente y la funcion

plt.title('Distribución F Snedecor')
plt.ylabel('Probabilidad')
plt.xlabel('X :Tasa de variabilidad entre Fabricas')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

#  **k. Suponga que la tasa de variabilidad sigue una distribució F con $(5,8)$ como parametros, calcule el valor de $k$  tal que $P (k < X < 2.55) = 0.42$**

```python
# **k**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(0, 8, 401)
F = stats.f(5,8)

y = stats.f(5,8).pdf(x)
a = 2.55 # Valor de la variable a calcular a la izquierda
b = F.cdf(a)
c = F.cdf(a) - 0.42
d = round(F.ppf(c),2) # valor critico pedido
e = F.cdf(d) # Probabilidad acumulada
x1 = np.linspace(0, d, 401)
x2 = np.linspace(d, a, 401)
x3 = np.linspace(a, 8, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,F.pdf(x1) , color="r", alpha = 0.3)
ax.fill_between(x2, 0,F.pdf(x2) , color="b", alpha =0.3)
ax.fill_between(x3, 0,F.pdf(x3) , color="y", alpha =0.3)

ax.text(d, 0.00001, d)
ax.text(a, 0.00001, a)

ax.text(a, 0.2, f'P(X>{a}) = {1-b:.4f}')
ax.text((a+d)/2, 0.1, f'P({d}=<X<={a})= {b-e:.4f}')
ax.text(d, 0.15, f'P(X<={d}) = {e:.4f}')

plt.title('Distribución F Snedecor')
plt.ylabel('Frecuencias')
plt.xlabel('X :Tasa de variabilidad entre Fabricas')
ax.grid() # le da mas elegancia al grafico
plt.show()

print(f'Area solicitada = {c:.4f}')

print(f'El valor critico solicitado es k ={d}')
print(f'P(X<={d})={e:.4f}')
print(f'P({d}=<X<={a})={b-e:.4f}')
print(f'P(X<={a})={b:.4f}')

```
