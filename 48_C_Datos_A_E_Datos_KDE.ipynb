# **212. Análisis Exploratorio de Datos - Kernel Density Estimation KDE**

# **a. Base de datos - Kernel Density Estimation KDE**
+ KDE - Es similar al histograma y nos proporciona una curva de densidad de prob. continua lo que permite una interpretacion mas facil


```python
#**b. Kernel Density Estimation KDE**
datos=pd.read_csv("/content/df_datos_LIMPIOS.csv",delimiter=',')
datos
```

#**b. Dos variables: Una cuantitativa y la otra cualitativa**
+ **De la variable Programa estudiaremos la variable Edad para la Categoria Psicología**

```python
PSIC = datos.loc[(datos['PROGRAMA']=='PSIC'),['PROGRAMA','EDAD']] # Seleccionamos Psicologia de la varible programa
PSIC
```

# **c. Construimos el Histograma para la data construida**

```python
sns.histplot(data = PSIC['EDAD'])
plt.title('Histograma de Programa Psicologia')
plt.show() # desplliege en pantalla el grafico
```

# **d. Construimos la curva de Densidad**

```python
# **d. Construimos la curva de Densidad**
```

```python
plt.rcParams['figure.figsize'] = (6,4) # Controlamos el tamaño de la figura
sns.histplot(data = PSIC['EDAD'],kde=True) # agregamos la curva de densidad
plt.title('Histograma de Programa Psicologia')
plt.show() # desplliege en pantalla el grafico
```

# **e. Construimos solo la curva y sombreamos el área debajo de la curva**

 + **La curva ilustra las edades mas comunes de nuestros estudiantes de sicologia**

```python
fig = plt.figure() # estructura de la figura un solo grafico
ax = fig.add_subplot(1,1,1) # # add_subplot es la figura creada

sns.kdeplot(data = PSIC['EDAD'],color = 'g', shade = True) # color green y area

ax.set_ylim(0,0.3) #limite de los ejes
ax.set_xlim(14,30)

ax.set_title('Grafico densidad de Programa Psicologia')
plt.show() # desplliege en pantalla el grafico
```

# **i. Ahora usamos todas las modalidades de la variable programa**

```python
plt.rcParams['figure.figsize'] = (6,6)
fig = plt.figure() # estructura de la figura un solo grafico
ax = fig.add_subplot(1,1,1) # # add_subplot es la figura creada

sns.kdeplot(data =datos, x = 'EDAD', hue = 'PROGRAMA', shade = True) # color green y area

ax.set_ylim(0,0.08) #limite de los ejes
ax.set_xlim(10,30)

ax.set_title('Grafico densidad de Programa Psicologia')
plt.show() # desplliege en pantalla el grafico
```

# **j. Ahora usamos todas las modalidades de la variable programa con la variable estatura**

# **k. Estudiando dos variables**
